## 14.0 セットアップ

```python
import numpy as np
from numpy.linalg import eigh
np.set_printoptions(precision=4, suppress=True, linewidth=120)

N = 3  # 与件（QCD との同定）
print("setup done. N =", N)
```

## 14.1 動力学原理 (a)：全結合系に RG の「スケール」はあるか

RG の本質は短距離自由度の積分消去（粗視化）である。だが本丸の作用は**全結合**
（結合＝相関行列の台＝全ペア、nb10）。全結合系には距離による粗視化の階層がない
（平均場が厳密になる極限）。「短距離」が定義されないため、素朴な実空間 RG は空回りする。
系サイズ `n`（粗視化スケールの代理）を変えても有効相関の構造（rank）が不変なことを確認する。

```python
print("系サイズ n を変えたときの相関行列 C の rank（粗視化スケール依存性）：")
for n in [50, 100, 200, 400]:
    th = np.linspace(0, 2*np.pi, n, endpoint=False)
    C = np.cos(2*(th[:, None] - th[None, :]))/(2*N)
    ev = np.sort(np.abs(eigh(C)[0]))[::-1]
    rank = int((ev > 1e-9*ev[0]).sum())
    print(f"  n={n:4d}: rank(C)={rank}, 上位2固有値={ev[:2].round(4)}, "
          f"残り最大={ev[2]:.2e}")
print()
print("→ どの n でも rank2。系サイズ（粗視化スケール）を変えても有効相関は rank2 のまま。")
print("  RG 流のパラメータ空間が rank2 部分空間に閉じる → 周期性を破る方向に流れない。")
```

## 14.2 動力学原理 (b)：N(μ) を走らせても周期性は破れるか

QCD の漸近自由性の類似で、`N`（または結合）をスケール `μ` 依存にする方向を検討する。
nb12 機構1 で確定した通り、`N` は振幅 `1/(2N)` にのみ入り角度構造（周期）には入らない。
よって `N(μ)` をどう走らせても、距離の角度依存（中心化後）は不変であることを確認する。

```python
print("N(μ) を変えたときの中心化距離の角度構造（nb12 機構1 の μ 依存版）：")
for Nmu in [3.0, 2.5, 4.0, 1.5]:
    phi = np.linspace(0.001, np.pi/2 - 0.001, 1000)
    d = -np.log(np.abs(np.cos(2*phi))/(2*Nmu))
    dc = d - d.mean()
    print(f"  N(μ)={Nmu}: 中心化距離の標準偏差={dc.std():.6f}")
print()
print("→ 全て同値。N(μ) は『振幅の再規格化』であって『周期構造の変更』ではない。")
print("  フーリエ台は cos2θ の {cos2kθ,sin2kθ} に限定（k*=4, nb02）、rank2 不変（nb05）。")
print("  N(μ) を走らせても周期性は破れない。")
```

**動力学原理の判定**：(a)(b) より、動力学原理は**周期性を破らない**。これは引き継ぎ書 C-3 の
懸念「本丸の結果と矛盾しないか」への回答＝**矛盾しない（整合的）**。だが裏を返せば、動力学原理は
**非コンパクト化も単一時間も出さない**。効きうるのは連続極限の取り方・次元の流れであって、目標の
時間／非コンパクト要素ではない（nb12 表「効く目標要素：連続極限・次元」と整合）。動力学原理は
符号原理 II（時間）や次元原理（空間）と**役割が異なり競合しない**補助原理に位置づくのが正しい。

## 14.3 非二体相互作用 (a)：与件内三体（cos2 の合成）は周期性を破るか

本丸 no-go の根因（nb05,06）：二体相関は rank2 カーネル
`cos2(θi−θj) = cos2θi·cos2θj + sin2θi·sin2θj`（加法定理）。固有空間が `{cos2θ, sin2θ}` の
2次元に限定され、コンパクト・周期的になる。三体相関 `E(θi,θj,θk)` で rank を上げれば根因を
破りうる。まず**与件内**（cos2 の合成で作る三体）が周期性を破るかを見る。

```python
m = 64
th = np.linspace(0, 2*np.pi, m, endpoint=False)

# 与件内三体：cos2(θi−θj)·cos2(θj−θk)（cos2 の合成、差ベース）
T1 = (np.cos(2*(th[:, None, None] - th[None, :, None]))
      * np.cos(2*(th[None, :, None] - th[None, None, :])))

# i 方向の全フーリエ成分（j,k で平均した振幅スペクトル）
F_i = np.zeros(m//2 + 1)
for j in range(m):
    for k in range(m):
        F_i += np.abs(np.fft.rfft(T1[:, j, k]))
nz = [int(k) for k in np.where(F_i > 1e-6)[0]]
print(f"与件内三体 cos2(i−j)cos2(j−k) の i 方向フーリエ台（振幅>1e-6 の次数）: {nz}")
print(f"  偶数次のみか（周期セクター内 {{cos2kθ,sin2kθ}}）: {all(k % 2 == 0 for k in nz)}")
print()
print("→ 与件内三体のフーリエ台は偶数次のみ。{cos2kθ,sin2kθ} の周期セクターに閉じる。")
print("  三体に拡張しても本丸 no-go（周期性不変）は破れない（加法定理の代数的帰結、堅牢）。")
```

## 14.4 非二体相互作用 (b)：周期性を破る三体は何を要求するか

では周期性を破るにはどんな三体が要るか。与件外の**非周期**関数形（θ を周期 2π で巻き戻さず
R 上で線形に使う）を入れた場合のフーリエ台を見る。これが奇数次・連続スペクトルを持てば
周期性を破るが、その代償（何を仮定したか）を厳密に読む。

```python
# 与件外・非周期三体：(θi−θj)(θj−θk)（R 上の多項式、周期2πで巻き戻さない）
T2 = ((th[:, None, None] - th[None, :, None])
      * (th[None, :, None] - th[None, None, :]))
F_i2 = np.zeros(m//2 + 1)
for j in range(m):
    for k in range(m):
        F_i2 += np.abs(np.fft.rfft(T2[:, j, k]))
F_i2 /= (m*m); F_i2 /= F_i2.max()
nz2 = [int(k) for k in np.where(F_i2 > 1e-3)[0]]
print(f"与件外非周期三体 (θi−θj)(θj−θk) の i 方向フーリエ台（振幅>1e-3）: {nz2[:12]} ...")
print(f"  奇数次 k=1,3,… を含むか（周期セクター外＝周期性を破る）: "
      f"{any(k % 2 == 1 for k in nz2)}")
print()
print("→ 非周期三体は奇数次を含み周期性を破る。だがその非周期性の正体は")
print("  『θ を R 上で巻き戻さず使う』＝設定空間を S¹ でなく R とみなすこと。")
print("  これは nb13 の被覆原理 II と同一の要請である。")
```

**非二体相互作用の判定**：
- **与件内**（cos2 の合成）に留まる限り、三体でもフーリエ台は偶数次の周期セクターに閉じ、
  本丸 no-go は破れない（14.3）。
- **周期性を破る**三体は、必ず**与件外の非周期関数形**を要し、その非周期性は「設定空間を
  普遍被覆 R とみなす」こと、すなわち**符号原理 II（nb13）に帰着**する（14.4）。

したがって C-4 は「ランク2 を破る独立した新原理」ではなく、**被覆原理 II を相互作用（多体）の
言葉で言い換えたもの**に近い。三体という形式が新しいだけで、周期性を破る源泉は II と同じ
（S¹ → R の非周期化）。独立な追加自由度を増やしていない点で、これは**朗報**である：
追加原理の地図が II に集約され、最小性が保たれる。
