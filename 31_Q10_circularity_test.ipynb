## 31.1 Tsirelson 飽和の標準幾何（照合の基準・文献事実）

CHSH 飽和の標準的な幾何を確認する（精読済み文献：Janas–Cuffaro–Janssen 1910.10688 §4、
量子チュートリアル 2511.09507 App.C）。これが Q10 の円と照合する**基準**になる。

1. **相関＝単位ベクトルの内積.** 測定設定を単位ベクトル `ê_a, ê_b` で表すと、量子相関は
   `E(a,b) = ê_a · ê_b = cos∠(a,b)`。相関の振幅は**単位ベクトルゆえ最大1**（完全相関）。
2. **飽和＝角度配置.** CHSH `S = E(a,b)+E(a,b')+E(a',b)−E(a',b')` が `2√2` を達成するのは、
   設定ベクトルが **45° 間隔**に並ぶとき（`cos45°=1/√2` が4項で `4·(1/√2)=2√2`）。
3. **上界 2√2 の起源＝非可換性.** `‖B²‖ ≤ 4 + ‖[a₁,a₂]‖·‖[b₁,b₂]‖ ≤ 4+4=8`、ゆえ `‖B‖≤2√2`
   （2511.09507 App.C eq.55）。上界そのものは**作用素の非可換性**から来る（nb30 Q0 が仮定した層）。

**Q10 にとっての含意（照合の枠組み）**：
- 標準幾何は「相関＝cos（単位円上の内積）」「飽和＝角度配置」の二つを言う。
- 本研究の特徴写像 `φ=(cos2θ,sin2θ)/√(2N)` は、まさに「相関＝円上の内積」を実現している
  （`φ(θ_i)·φ(θ_j) = cos2(θ_i−θ_j)/(2N) = C`）。ここが Q10 の主張の幾何的核心。
- ただし本研究の円は**半径 1/√(2N)**（単位円でない）。標準の量子相関を `1/(2N)` でスケールした形
  ＝N=3 の color-singlet qutrit 相関 `cos2θ/6`（A3・E-4 の設定）に対応。

```python
import sympy as sp
theta_i, theta_j = sp.symbols("theta_i theta_j", real=True)
N = sp.symbols("N", positive=True)

# 特徴写像の内積が相関 C を再現することを確認（Q10 主張の幾何的核心）
phi_i = sp.Matrix([sp.cos(2*theta_i), sp.sin(2*theta_i)])/sp.sqrt(2*N)
phi_j = sp.Matrix([sp.cos(2*theta_j), sp.sin(2*theta_j)])/sp.sqrt(2*N)
inner = sp.simplify((phi_i.T @ phi_j)[0])
C = sp.cos(2*(theta_i - theta_j))/(2*N)
print("φ(θi)·φ(θj) =", inner)
print("C(θi,θj)    =", sp.simplify(C))
print("一致：", sp.simplify(inner - C) == 0)
print()
print("⇒ 特徴写像の内積が相関を再現。『相関＝円上の内積』は Tsirelson 幾何と同じ構造。")
print("  ただし円の半径は 1/√(2N)（標準の単位円を 1/√(2N) でスケール）。")
```

## 31.2 二起源分解を標準幾何に照合する（判定の本体）

nb22b §22b.3 で `|u|² = R(N)²·|ê|²` と分解した。これを §31.1 の標準幾何の二要素
（相関＝cos / 飽和＝角度配置）に照合し、**飽和の何が起源 (a) で何が起源 (b) か**を確定する。

### 31.2.1 起源 (a)：相関が円（cos）に乗ること＝飽和の幾何的構造

標準幾何の第一要素「相関＝単位ベクトルの内積 cos」は、本研究では `|ê|²=cos²+sin²=1`
（方向部分の恒等式）が担う。これは **N に依らない＝与件の振幅と独立**（22b.3 起源 a）。

**飽和の幾何的構造との対応**：CHSH が 2√2 を達成する角度配置（45°間隔）は、相関が
**cos の形**（角度の内積）であることに全面的に依存する。cos でなければ 45°配置で 2√2 にならない。
そして「相関が cos である」＝「特徴像が円である」＝起源 (a)。

→ **飽和の幾何的構造（cos 形・円・45°配置で最大化が起こりうること）は起源 (a) から、
与件の振幅を仮定せずに出る。** これは循環でない（G1 PASS）。

```python
# 起源(a)の検証：相関が cos 形なら CHSH の最大値が 2√2·(振幅) になることを確認。
# 単位円(振幅1)で 45°配置 → 2√2。本研究の円(振幅 1/(2N)) → 2√2/(2N)。
import numpy as np

def chsh_from_circle(radius, angles):
    # angles = (a, a', b, b') の設定角。相関 E = radius² · cos2(Δθ)... ではなく
    # 標準 CHSH 幾何に合わせ E(x,y)=radius²·cos(θx-θy) の符号付き和を最大化。
    a, ap, b, bp = angles
    def E(x, y): return radius**2 * np.cos(x - y)
    return E(a,b) + E(a,bp) + E(ap,b) - E(ap,bp)

# 45°間隔の最適配置（標準）
opt = (0, np.pi/2, np.pi/4, -np.pi/4)
for r, label in [(1.0, "単位円（標準量子相関）"), (1/np.sqrt(2*3), "本研究の円 N=3（振幅1/6）")]:
    val = chsh_from_circle(r, opt)
    print(f"{label}: CHSH最適値 = {val:.4f}  (= 2√2·radius² = {2*np.sqrt(2)*r**2:.4f})")
print()
print("⇒ 飽和の『構造』（45°配置で最大化＝cos形ゆえ）は半径に依らず起こる＝起源(a)。")
print("  飽和の『数値』は radius²=1/(2N) でスケール＝起源(b)（次節）。")
```

### 31.2.2 起源 (b)：飽和の数値（スケール）＝振幅 R(N)²=1/(2N)＝A3 依存

標準幾何の飽和値は **2√2**（単位円・振幅1）。本研究の円は振幅 `1/(2N)` ゆえ、CHSH 最適値は
`2√2 · 1/(2N) = √2/N`。この**数値**は振幅 `R(N)²=1/(2N)` に比例し、`N` を含む（22b.3 起源 b）。

**A3 依存の確定**：`N` は引き継ぎ書 A3（N=3 の QCD 同定）で**外部入力**として固定される量。
よって飽和の**数値**（円の半径＝相関振幅）は与件 cos2θ の関数形だけからは決まらず、**A3 を経由**する。

**重要な区別（D-28 / nb30 §1 の L-b / L-c）**：
- ここで言う「飽和の数値」は、相関の**絶対スケール** `1/(2N)`。これは A3 依存。
- 一方、CHSH が古典束(2)でなく量子束(2√2)を**飽和する**という事実（量子集合の境界に乗ること、
  すなわち L-b）は、相関が cos 形であること＝起源 (a) と、上界が非可換性で 2√2 に決まること
  （§31.1 の第3要素）から来る。**スケールを除いた飽和の質（量子境界に乗ること）は起源 (a) 側**。
- N はスケール（振幅）を決めるだけで、「量子境界に乗るか否か」は決めない（N をどう変えても
  cos 形なら同じく量子境界に乗る）。

```python
# 起源(b)の検証：飽和の数値は N に依存（A3経由）、飽和の質（境界に乗るか）は N 非依存。
print("飽和の数値（CHSH最適値）の N 依存：")
for Nval in [2, 3, 5, 10]:
    val = 2*np.sqrt(2) / (2*Nval)
    print(f"  N={Nval:2d}: CHSH最適 = 2√2/(2N) = {val:.4f}")
print("  → 数値は N に依存（A3 経由）＝起源(b)。")
print()
# 飽和の質：規格化相関（振幅で割る）で見れば、どの N でも同じ「量子境界に乗る」配置。
print("飽和の質（振幅で規格化した相関の角度配置）の N 依存：")
for Nval in [2, 3, 5]:
    # 規格化相関 E/振幅² = cos(角度) は N に依らず同じ → 同じ 45°配置で同じ「境界に乗る」
    normalized_opt = 2*np.sqrt(2)  # 規格化後はどの N でも 2√2
    print(f"  N={Nval}: 規格化 CHSH = {normalized_opt:.4f}（N 非依存）")
print("  → 質（量子境界=2√2 に乗ること）は N 非依存＝起源(a)。")
```

## 31.3 飽和の達成を内部集合上で言う（nb30b 規約 E-1 の適用）

§31.2 は「飽和が起こりうる」を幾何的に示したが、nb30b 規約 E-1 は「**達成（最大値の存在）は
内部集合の上で言え**」と要求する。Q10 の飽和達成がこれを満たすか確認する。

- 設定の角度配置の集合（45°間隔の4点）は**超有限格子 `θ_i=i·ε` 上の内部集合**の点として実現される
  （nb22b §22b.2.1：像 \*S¹ は内部集合）。
- 最適配置（CHSH を最大化する4点）は内部集合上の**超有限個の点**にわたる最大化で、内部的に達成元が
  存在する。`st` を取って標準の飽和値 `√2/N`（N=3 で `√2/3`）を得る。
- **Q10 は最大化を円の幾何（45°配置）に帰着させるので、外部集合上の上限（達成されない sup）の罠
  （E-1 が警告する外部性）を回避する。** これは nb30 で Q10 を Q9 より上位に置いた理由の一つ。

```python
# 内部集合上の達成：超有限格子の点として最適配置が存在することの標準的傍証。
# （*R 上では超有限和の最大化。ここでは標準格子で「格子を細かくすれば最適配置に到達」を傍証。）
best = -np.inf
for M in [8, 90, 360, 3600]:   # 格子点数（*R では超有限 M に対応）
    grid = np.arange(M) * (2*np.pi/M)   # θ_i = i·ε, ε=2π/M
    # 4設定を格子から選ぶ最大化は重いので、最適配置近傍の格子点で評価
    r = 1/np.sqrt(2*3)
    def E(x,y): return r**2*np.cos(x-y)
    # 最適 (0, π/2, π/4, -π/4) に最も近い格子点
    pts = [min(grid, key=lambda g: abs(((g-t+np.pi)%(2*np.pi))-np.pi))
           for t in (0, np.pi/2, np.pi/4, -np.pi/4)]
    a,ap,b,bp = pts
    val = E(a,b)+E(a,bp)+E(ap,b)-E(ap,bp)
    best = max(best, val)
    print(f"M={M:5d} (ε=2π/M): 格子上 CHSH = {val:.5f}")
print(f"\n格子を細かく → √2/3 = {np.sqrt(2)/3:.5f} に収束。")
print("⇒ *R 上では超有限 M で内部集合上に達成元が存在。st で標準値 √2/3。E-1 を満たす。")
```

## 31.4 nb30 rubric への差し戻しと Q10 の最終ラベル（二層判定）

§31.2–31.3 の結果を nb30 の rubric（G1・G2・G3'・G4・G5・G6）に戻し、Q10 を**二層に分けて**
判定する。これが本ノートの結論。

```python
from dataclasses import dataclass, field
from typing import Dict
PASS, FAIL, UNK = "PASS", "FAIL", "UNK"
GATES = ("G1","G2","G3p","G4","G5","G6")

@dataclass
class Layer:
    name: str
    gates: Dict[str,str]
    note: str
    def verdict(self):
        v=[self.gates.get(g,UNK) for g in GATES]
        if FAIL in v: return "KILL"
        if UNK in v: return "HOLD"
        return "GO"

layers = []

# 層 a：飽和の幾何的構造（質）
layers.append(Layer(
    "Q10a 飽和の幾何的構造（質・cos形・量子境界に乗ること）",
    dict(G1=PASS, G2=PASS, G3p=PASS, G4=PASS, G5=PASS, G6=PASS),
    "起源(a)=|ê|²=1（恒等式・与件振幅と独立）。相関がcos形＝円ゆえ45°配置で量子境界2√2に乗る。"
    "G1非循環(飽和値を仮定せず円の幾何から)、G2量子非仮定、G3'cos2θ整合、G4 cos²+sin²=1は代数恒等(本数合わせでない)、"
    "G5 古典束2でなく量子束に乗ることを選ぶ、G6 |ê|²=1はcos2θの振幅より上流(振幅に依らない)。"))

# 層 b：飽和の数値（スケール）
layers.append(Layer(
    "Q10b 飽和の数値（スケール・半径1/√(2N)）",
    dict(G1=PASS, G2=PASS, G3p=PASS, G4=UNK, G5=FAIL, G6=FAIL),
    "起源(b)=振幅R(N)²=1/(2N)、N は A3(QCD同定)で外部固定。"
    "G5 FAIL: 数値は A3 を入力に取って初めて決まり、与件cos2θだけでは一意化しない。"
    "G6 FAIL: スケールの決定は A3 という cos2θ と別の外部入力に依存＝上流原理になっていない。"
    "→ 飽和の『数値』は E-2 の上流原理としては出ない（A3 依存）。"))

print("="*92)
print(f"{'層':<46}{'G1':>6}{'G2':>6}{'G3⁰':>6}{'G4':>6}{'G5':>6}{'G6':>6}  判定")
print("="*92)
for L in layers:
    g=L.gates
    print(f"{L.name:<46}{g['G1']:>6}{g['G2']:>6}{g['G3p']:>6}{g['G4']:>6}{g['G5']:>6}{g['G6']:>6}  {L.verdict()}")
print("="*92)
for L in layers:
    print(f"\n[{L.verdict()}] {L.name}")
    import textwrap
    for ln in textwrap.wrap(L.note, 88): print("   ", ln)
```
