**結論の先取り**：cos2θ から空間構造が出るという*着想*は Karlsson と共有し、新規ではない。本研究の

## 23.0 セットアップ

```python
import numpy as np
import sympy as sp
np.set_printoptions(precision=5, suppress=True, linewidth=120)
N = 3
print("setup done. N =", N)
```

## 23.1 一致点：cos2φ の直交軸分解（Karlsson 式10 ＝ 本研究の補題）

Karlsson 式(10) は `e^{2iφ}=cos2φ+i sin2φ` の実部・虚部分解と、π周期が φ∈{0,π/4} の直交軸に
対応することを述べる。本研究 nb22 の特徴写像 `u=(cos2θ,sin2θ)/√(2N)` と同じ2成分構造である。
この着想は**共有されており新規性はない**。

```python
phi = sp.symbols("phi", real=True)
karlsson = sp.cos(2*phi) + sp.I*sp.sin(2*phi)
print(f"Karlsson 式10: e^(2iφ) の実部={sp.re(sp.cos(2*phi))}, 虚部=sin2φ")
print("本研究 nb22 : 特徴写像 u=(cos2θ, sin2θ)/√(2N)、同じ2成分直交構造。")
print("→【一致・新規性なし】cos2φ の (cos2φ,sin2φ) 分解、π周期→直交軸 φ∈{0,π/4}。")
```

## 23.2 差分1：Karlsson は距離核・MDS を構成しない

本研究は cos2θ を距離 `d=−log|C|` に変換し MDS で計量を出す。Karlsson は cos2φ を相補性
（スカラー積）の表現とするにとどまり、距離核 `−log|C|` も MDS による計量復元もない。

```python
m = 100
th = np.linspace(0, 2*np.pi, m, endpoint=False)
C = np.cos(2*(th[:, None] - th[None, :]))/(2*N)
d = -np.log(np.abs(C)); d[~np.isfinite(d)] = 0.0; np.fill_diagonal(d, 0.0)
J = np.eye(m) - np.ones((m, m))/m
B = -0.5*J@(d**2)@J
ev = np.sort(np.linalg.eigvalsh(0.5*(B + B.T)))[::-1]
print(f"本研究：d=−log|C| → MDS 固有値比 {(ev[:4]/ev[0]).round(3)} → S¹ 計量を構成。")
print("Karlsson：cos2φ をスカラー積の表現とするのみ。距離核も MDS もなし。")
print("→【差分1】計量の明示的・構成的導出（d=−log|C|→MDS）は本研究固有。")
```

## 23.3 差分2（最重要）：Karlsson の周期倍化 vs 本研究の被覆II

最も混同しやすい点。Karlsson 式(11) `(a·b)²=(1+ã·b̃)/2` は φ→2φ で周期を π→2π にする（2d→3d
双対）。本研究の被覆II（nb13）は S¹→R で周期を π→∞ にする（普遍被覆）。**両者とも「周期を変える」
が行き先が決定的に違う**：Karlsson は有限倍化（2π、コンパクト S¹ のまま、空間次元の双対）、本研究は
非周期化（R、非コンパクト、時間の創発）。似て見えて別物であることを明示する。

```python
print("Karlsson 式11: φ→2φ で周期 π→2π。行き先は 2π周期＝まだコンパクト S¹（空間内）。")
print("本研究 被覆II : S¹→R で周期 π→∞。行き先は普遍被覆 R＝非コンパクト（時間）。")
print()
print("  操作         | 周期の行き先 | 位相      | 出てくるもの")
print("  -------------|-------------|-----------|-------------------")
print("  Karlsson倍化 | π → 2π      | コンパクト | 2d↔3d 相関の双対")
print("  本研究 被覆II | π → ∞       | 非コンパクト| 単一・非コンパクト時間")
print()
print("→【差分2・最重要】被覆II は Karlsson の倍化の焼き直しではない。")
print("   倍化は空間内で周期を有限に変える操作、被覆IIは時間Rを生む位相的操作。")
print("   両者は『周期を変える』点だけが表面的に似ており、本質は別。")
```

## 23.4 差分3：次元3について結論が逆向き

Karlsson は「spin1/2 が D=4 の唯一の候補」とし、式(11)の双対で**次元3を肯定的に主張**する。本研究は
nb17/nb21 で**次元3は cos2θ と独立・外部入力**と否定的に確定した。同じ cos2θ から出発して結論が逆。

```python
print("Karlsson：spin1/2 が D=4 の唯一候補 → 次元3を肯定的に主張（式11の倍化双対）。")
print("本研究  ：次元3は cos2θ と独立、外部入力 → 否定的に確定（nb17, nb21）。")
print()
print("→【差分3】本研究は Karlsson の『spin1/2→3d』に反例的含意を持つ：")
print("   nb21 で cos2θ の登攀（軸I）に次元は現れない。Karlsson の倍化双対は")
print("   『次元を生成する』のでなく『2dと3dの相関形を関係づける』にとどまる、")
print("   と本研究の立場からは読める。次元3の起源は依然未解決（cos2θ の外）。")
```

## 23.6 サマリー（established）

### 共有されており新規性のない点（正直に記録）

1. **cos2θ の直交分解**（cos2θ, sin2θ の2成分、π周期→直交軸 φ∈{0,π/4}）は Karlsson 式(10)と一致。
2. **cos2θ（spin/photon 相関）から空間の向き・直交軸が出るという着想**は Karlsson と共有。

### 本研究固有の寄与（Karlsson が定性段階で止めた部分の構成的・定量的な詰め）

1. **計量の構成的導出**：`d=−log|C|`→MDS で S¹/T³ の計量を明示的に復元（差分1）。Karlsson に距離核なし。
2. **被覆IIによる非コンパクト時間**：S¹→R の普遍被覆（π→∞）。Karlsson の周期倍化（π→2π、コンパクト
   保持、空間次元の双対）とは行き先も役割も別物（差分2、最重要）。
3. **次元3の否定的確定**：cos2θ と独立・外部入力（nb17,21）。Karlsson の「spin1/2→3d」肯定的主張と
   逆向きで、反例的含意を持つ（差分3）。
4. **no-go の解析証明・標準部写像 st・質量ギャップ**：Karlsson に対応物なし（差分4）。

### 新規性の正確な要約

「ベル相関 cos2θ から空間構造が出る」という*着想*は新しくない（Karlsson 2018 が先行）。本研究の
寄与は、その着想を **(i) 計量として構成的に導出し、(ii) no-go を解析的に証明し、(iii) 時間を被覆IIで
非コンパクト化し、(iv) 次元3を否定的に確定し、(v) st 測定論に載せた** 点にある。誇張は避け、
着想の先行性を認めた上で構成・定量・否定的結果の寄与を主張する。
