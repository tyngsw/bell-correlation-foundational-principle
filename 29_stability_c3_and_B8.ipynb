## 29.1 まず枠の与件だけで：作用 `S=−βΣe^{−d}` は次元を含むか

C3 が力学を密輸せず d を選ぶには、**枠が既に持っている資源**＝普遍作用
`S = −β Σ e^{−d}`（README/A 系）だけで d 依存性が出ねばならない。
ところが `d = −log|C|` より `e^{−d} = |C|`。つまり `S = −β Σ |C|`。検算する。


```python

import numpy as np
N=3
def phi(t): return np.array([np.cos(2*t),np.sin(2*t)])/np.sqrt(2*N)
th=np.linspace(0,2*np.pi,40,endpoint=False)
F=np.array([phi(t) for t in th])
C=F@F.T                          # 相関行列
d=-np.log(np.abs(C)+1e-12)       # 距離
S=-(np.exp(-d)).sum()            # S=-βΣe^{-d}, β=1
print(f"Σe^(-d) と Σ|C| の一致: {np.allclose(np.exp(-d), np.abs(C))}")
print(f"S = -βΣe^(-d) = -βΣ|C| (β=1) = {S:.3f}")
print()
print("→ 普遍作用は相関の総和に等しく、*埋め込み次元 d を含まない*。")
print("  枠の与件だけでは作用は次元に対して盲目＝C3 は枠内資源で d を選べない。")
print("  d 依存性を入れるには運動項(ラプラシアン on R^d)等を *外から* 足すしかない（G1 の山場）。")
```

## 29.2 力学を入れたとして：安定性は d=3 を一意化するか

仮に外部から運動項を入れ、Ehrenfest/Case 型の安定性（量子原子が崩壊しない・束縛軌道が安定）を
論じても、得られるのは **d に関する不等式**であって一意化ではない、ことを確認する。


```python

print("ラプラシアンの Green 関数から来る Coulomb 型 V(r) と量子安定性（Case 1950 型）：")
print(f"{'d':>3}  {'V(r)~':>8}  {'量子原子が崩壊しない':>18}")
for dd in [1,2,3,4,5]:
    if dd==2: pot="log r"
    else: pot=f"r^{-(dd-2)}" if dd!=2 else "log r"
    stable = dd<4   # 崩壊回避は d<4
    print(f"{dd:>3}  {pot:>8}  {str(stable):>18}")
print()
print("→ 『安定』は d<4 という *不等式*。d=1,2,3 を同格に許し、d=3 を一意化しない。")
print("  d=3 を一意に選ぶには Bertrand(閉軌道は d=3 の逆二乗のみ)や Huygens(無反響波伝播)など")
print("  *さらに別の外部原理* が要る。C3 は外部力学を二重に密輸する（G1 二重違反）。")
```

## 29.3 C3 の判定：G1 二重密輸で不成立

- **29.2**：外部から力学（運動項）を入れても、安定性は **d<4 の不等式**で **d=3 を一意化しない**。
  d=3 のクリーンな選択には Bertrand／Huygens 等の**さらなる外部原理**が要る。

**∴ C3 は次元3の上流原理にならない。** C3 が d を選ぶには (i) 運動項という外部力学と
