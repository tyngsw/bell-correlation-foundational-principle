# nb24 — 実空間3次元の上流原理：候補リストと D-10 免除の事前判定

**位置づけ（E-1, F節）.** 第一段階で残った唯一の核心的空白は「実空間が3次元であること」の上流原理である。
引き継ぎ書で次の二点が**否定的に確定**している：

- **B-2 / nb15, nb17**：与件 `cos2θ` ＋ 測定論 `st` からは独立設定本数 `p`（→空間次元）は予言できない。
  枠内では p=1,2,3,… が同格に無矛盾。次元は**外部入力**。
- **B-7 / nb21**：第二段階の登攀（P0→A1等質性→A2量子相関→A3 N=3）の**どの段にも次元3は現れない**。
  次元3は `cos2θ` と**直交な独立軸**（D-24）。

したがって E-1 は「`cos2θ` から次元を出す」のではなく、「`cos2θ` とは**別の最小原理**から
次元3を登る道があるか」を問う。本ノートはその研究（nb25 以降）の**着手前の安全装置**であり、
計算ノートではない。目的は次の二つ（F節の指示そのまま）：

1. 次元3を出しうる**候補原理のリストアップ**（網羅と分類）。
2. 各候補が **D-10（値を後付けする偽の原理 vs 値を予言する本物の原理）をどう免れるか**の事前判定。

空振り（全候補が落ちる）も nb15 / nb17 と同様に**否定的確定として価値がある**（D-15, D-18, D-24）。
本ノートの出力は「どの候補に計算資源を投じてよいか」の **go / hold / kill** リストである。


## 0. 適用する鉄の教訓（判定の根拠規則）

本ノートの判定は気分でなく、引き継ぎ書 C 節の以下を**機械的に**適用する：

- **D-10**（最重要）：パラメータ先決め＋意味の後付けは偽の原理。本物は値を**予言**する。
  → 次元3を出力に持つ候補は、「3 を入力のどこかに忍ばせていないか」を最優先で疑う。
- **D-13**：数の一致は原理でない（QCD の3、4N の4、空間の3 …）。起源を分離せよ。
- **D-17**：数の一致が**群・代数の一致**とは限らない（D-13 の構造版）。
  → 「○○が3つあるから空間3次元」は本数の対応にすぎない。群・代数レベルの必然性を要求する。
- **D-14 / D-22**：目標要素を与件の内外で分類し、複合した疑いは要素に分解せよ。
- **D-26**：似た操作は**行き先**（位相・次元・コンパクト性）で区別せよ。
- **A-8**：`st` は次元を縛らない（上流が st より上）。**st 系の原理は E-1 の解にならない**（除外規則）。


## 1. 判定 rubric（先に固定する）

候補ごとに恣意性が入らぬよう、**5 つの関門**を先に定義し、順に適用する。
いずれかで FAIL すれば、その時点で **KILL**（または条件付きで HOLD）。

| 関門 | 名称 | 問い | FAIL 条件 |
|---|---|---|---|
| **G1** | **非循環（D-10）** | 出力 d=3 が、入力・設定・台の選択に「3」として埋め込まれていないか | 3 を仮定して 3 を出している（後付け） |
| **G2** | **st 非依存（A-8）** | 原理が `st` だけに依拠していないか | 本質が st に還元される（次元を縛れない） |
| **G3** | **cos2θ 独立（B-7）** | `cos2θ` の再導出に化けていないか／逆に cos2θ を密輸していないか | cos2θ かその登攀段に等価 |
| **G4** | **群・代数水準（D-17）** | 「3」が本数の一致でなく**必然**として出るか | 数合わせ（本数対応のみ）で群・代数の必然性なし |
| **G5** | **予言力（D-10 正面）** | d=3 を**一意に**選び、d≠3 を**排除**するか | d=2 も d=4 も同格に許す（一意化できない） |

**合否ラベル**：
- **GO** … G1–G5 を原理的に通過可能（少なくとも FAIL が見えない）。計算で詰める価値あり。
- **HOLD** … 一部関門が未決定。先に概念的詰めが要る（計算前に潰すべき曖昧さ）。
- **KILL** … いずれかで原理的に FAIL。再挑戦は時間の無駄（B 節同様の否定的確定候補）。

注意：GO は「正しい」ではなく「**まだ落ちていない**」の意。D-21（自分の結論を弱める検証を率先せよ）に従い、
GO 判定にも反証チェック（CHSH 類似の落とし所）を併記する。


## 2. 候補原理のリストアップ

次元3を出しうる機構を、**起源の種類**で分類して列挙する。nb15 で否定された5機構（B-3）は
「数の一致」型なので原則 D-13/G4 で再排除されるが、**精密版**（D-15）として再掲し落ち方を記録する。
新規候補は nb15 の射程外（=本数合わせでない構造的要請）を狙う。

分類：
- **(I) 代数・非可換性由来** … 相関の構造が要求する最小次元
- **(II) 安定性・力学由来** … 力学/くりこみ/安定性が次元を選ぶ
- **(III) 情報・確率構造由来** … 確率多面体・文脈依存性の次元
- **(IV) 位相・埋め込み由来** … 台を埋める最小幾何
- **(V) 人間原理・選択効果** … 観測者の存在条件（最後の砦、原則 D-10 で弱い）
- **(0) nb15 既否定（精密版で再掲）** … 数の一致型

各候補を rubric にかける。


```python

# 候補原理を構造化データとして定義し、rubric G1-G5 を適用して go/hold/kill を機械判定する。
# 判定は「原理的に FAIL が見えるか」の事前評価であり、計算結果ではない（本ノートは安全装置）。

from dataclasses import dataclass, field
from typing import Dict, List

PASS, FAIL, UNK = "PASS", "FAIL", "UNK"  # UNK = 未決定（HOLD要因）

@dataclass
class Candidate:
    cid: str
    cls: str          # 分類 I-V, 0
    name: str
    idea: str         # 一行アイデア
    gates: Dict[str, str] = field(default_factory=dict)  # G1..G5 -> PASS/FAIL/UNK
    note: str = ""

    def verdict(self) -> str:
        vals = [self.gates.get(g, UNK) for g in ("G1","G2","G3","G4","G5")]
        if FAIL in vals:
            return "KILL"
        if UNK in vals:
            return "HOLD"
        return "GO"

cands: List[Candidate] = []
```

## 3. 判定結果の読み（散文）

機械判定の要点を引き継ぎ書の論理に照らして整理する。

**KILL 群（再挑戦不要・否定的確定の候補）.**
nb15 既否定の C0a–C0c はいずれも G3/G4/G5 で落ちる——「3」が内部対称性（色・スピン）や数合わせ由来で、
**実空間次元と別物**（D-13, D-17）。C7（台を埋める最小幾何）は B-2 そのもので、3 軸を立てた結果の 3 を
出力と取り違える循環（G1）。C9（人間原理）は次元3を仮定して逆算する後付けの極北で、D-10 と正面衝突。
C5（相関多面体）は E-1 ではなく **E-2（量子相関の上流）側**に属し、cos2θ へ化ける（G3, B-7）。
C6（文脈依存の最小次元）の「3」は Hilbert 次元であって実空間でない——魅力的だが D-13 の罠が濃い。
C2（分割代数）は R,C,H,O が同格で d=3 を一意化できない（G5、B-2 の代数版）。

**HOLD 群（計算前に概念を詰めるべき）.**
4 件すべてが少なくとも一つの関門で UNK を残し、機械判定は全件 HOLD（GO ゼロ）となった。rubric は「UNK が一つでも残れば HOLD」と定めてあり（D-18：肯定・否定を同じ精度で扱う）、GO（全関門 PASS）に到達した候補は現時点で**存在しない**——これ自体が「次元3は容易には予言できない」という B-2 の追認である。HOLD 群は**有望度で二層**に分かれる。

*HOLD 上層（最有望・残る UNK が一つ）.* C8（絡み目の非自明性）と C3（安定性）。C1（非可換性の最小次元）と C4（くりこみ/臨界次元）。C1 は SU(2)~SO(3) 経由になりがちで、それは
**内部対称（スピン）であって空間ではない**（B-4：SO(3) は T³ の等長でない）——「非可換→3」が群の一致に
留まる危険（D-17）。C4 は nb19 の立方異方性（A-5）の延長で**枠内資源で動け密輸が少ない**点が最大の魅力。
ただし臨界次元論は通常 d を入力にする。**d を出力化できるか**が生死を分ける（G1）。E-5 と連結。

*HOLD 上層の中核* は C3 と C8 で、両者に共通する強みは **G5＝d を一意化する力**（PASS）。残る UNK は両者とも **G4（群・代数水準）** ただ一つで、これは「E-1 が群の必然まで要求するか」という**基準側の未決**であり候補の欠陥ではない。基準を緩めれば GO に最も近い。
- C3：逆二乗則の束縛安定が d=3 限定という Ehrenfest 型。d を一意化するが、**力学（作用・ポテンシャル）の導入で
  外部入力を密輸**する恐れが残る（G1 を計算段階で再審査）。本枠は今「設定空間の台」までしか持たない。
- C8：**「絡み目が非自明なのは厳密に d=3 のみ」**（d=2 では交差不可避、d≥4 では全て解ける）。これは
  位相の必然であり、cos2θ を密輸せず（G3）、本数合わせでもない（G4 寄り）。最有望の新規候補。
  ただし「entanglement の絡み」を**比喩でなく相関構造から定義**できるかが HOLD 的な未決要素。


## 4. D-10 免除の事前判定（本ノートの主目的）

F節が要求した「各候補が D-10 をどう免れるか」を、生き残り候補（GO/HOLD）について明示する。
D-10 免除とは「**次元3を入力に忍ばせず、原理から d=3 を一意に予言する**」こと。G1（非循環）と
G5（予言力）の合格がその実質である。

- **C8（絡み目, HOLD上層）— D-10 免除の筋が最も明快.**
  入力は「相関がほどけない（位相的に非自明な絡みを持つ）」という**定性的**要請のみ。出力 d=3 は
  「非自明な絡み目が存在する次元は厳密に3」という**数学的定理**から出る。3 を入力に書いていない（G1）、
  d=2 も d≥4 も排除される（G5）。**免除条件**：相関の「絡み」を `cos2θ`/MDS の構造から内在的に定義し、
  それが比喩でなく位相不変量（例：linking number 類似）として立つことを示す。これが nb25 の第一目標。

- **C4（臨界次元, HOLD）— 免除は「d の出力化」に懸かる.**
  nb19 の異方性展開は d を入力に固定して係数を見ている。D-10 を免れるには、**異方性の消失（連続回転対称の
  回復）を d の関数として書き、d=3 で特異/最適になる**ことを示す必要がある。それができれば 3 は予言、
