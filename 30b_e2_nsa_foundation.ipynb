## 1. 舞台の宣言：基礎体を \*R に取る

**規約 F-1（基礎体）.** 本研究の基礎体を超実数体 `\*R`（標準実数体 R の適切に飽和した超冪拡大、
少なくとも ℵ₁-saturated）とする。標準実数 R は `\*R` の部分体として埋め込まれ、`st: \*R_fin → R`
（有限超実数からその標準部への写像）が定義される。無限小 `ε>0`（`ε≈0`, `ε≠0`）を一つ固定する。

**規約 F-2（移行原理）.** 標準実数体 R で成り立つ一階の命題は `\*R` でもそのまま成り立つ（transfer）。
したがって R 上で確立済みの結果（nb22 の no-go の P1・P2・P3 を含む）は `\*R` 上の対応する命題として
**自動的に持ち上がる**。持ち上げた対象には `\*` を付して標準対象と区別する（例：`\*S¹`、`\*cos`）。

**規約 F-3（内部性）.** `\*R` 上で扱う集合・関数は原則として**内部的**（internal）とする。内部集合には
超有限性・移行原理が使え、最大値の存在などの議論が標準のコンパクト性の代わりに使える。**外部的**
（external）な対象（例：標準部の集合 R 自体、単子 monad）が現れる箇所は明示する。これは E-2 の
飽和（最大値の達成）を論じるとき決定的になる（§5、内部 vs 外部の罠）。

### 1.1 設定（コード規約：記号の対応表のみ。計算はしない）

```python
# 本ノートは規約ノートであり数値計算をしない。記号と層の対応を構造化して記録するのみ。
# （nb24 が Candidate を構造化したのと同じ「機械可読な規約」の精神。）

from dataclasses import dataclass, field
from typing import Optional

# st 発動の前後を表すラベル
PRE_ST   = "pre-st (内部・*R 上)"     # st を取る前：超実数の内部対象
POST_ST  = "post-st (標準・R 上)"     # st を取った後：標準実数対象
ALGEBRAIC = "algebraic (st 不要)"      # 代数的命題で st の前後を問わない（移行で両層に成立）

@dataclass
class Construct:
    name: str
    layer: str            # PRE_ST / POST_ST / ALGEBRAIC
    star_object: str      # *R 上での身分
    st_role: str          # st がこの構成にどう関与するか
    note: str = ""

constructs = []
print("nb30b: 規約ノート。構成物の層と st 発動点を地図化する。")
```

## 2. 各構成物の \*R 上での身分（地図の本体）

E-2 までに現れる主要構成物を一つずつ、`\*R` 上のどの対象か・st がどう関与するかで規定する。
**θ が標準実数に限るのか超実数まで動くのか**を各構成で明示する（飽和の議論に効く）。

### 2.1 与件 cos2θ と特徴写像

- **第一原理 `E(θ)=cos2θ/(2N)`**：移行原理により `\*cos` として `\*R` 上の超実数値関数に持ち上がる。
  θ は超実数まで動く（超有限格子 `θ_i=i·ε`, nb08）。標準的な cos2θ は `st` を取った像。
- **特徴写像 `u(θ)=(\*cos2θ, \*sin2θ)/√(2N)`**：`\*R²` への内部写像。像は `\*R²` 内の**超実数円 `\*S¹`**
  （半径 `1/√(2N)`）。標準円 S¹ は `st(\*S¹)`。
- **層の判定**：cos2θ・特徴写像・円は**代数的**（加法定理・`cos²+sin²=1` は移行で両層に成立）。
  ただし「円である」を**標準位相空間 S¹ として**読むには st が要る（pre-st では `\*S¹`）。

```python
constructs.append(Construct(
    name="第一原理 cos2θ/(2N)",
    layer=ALGEBRAIC,
    star_object="*cos2θ/(2N) : 超実数値関数（θ は超有限格子上を動く）",
    st_role="標準 cos2θ は st(*cos2θ)。関数形は移行で両層に成立。",
    note="nb08 の超有限格子 θ_i=i·ε が定義域。"))

constructs.append(Construct(
    name="特徴写像 u(θ)",
    layer=ALGEBRAIC,
    star_object="u: 内部写像 → *R²、像は超実数円 *S¹（半径 1/√(2N)）",
    st_role="標準円 S¹ = st(*S¹)。rank2・定ノルムは代数的（移行）。",
    note="nb22 の P1(rank2)・P2(定ノルム)はこの写像の代数的性質。微分を使わない。"))
```

### 2.2 距離核と MDS

- **距離核 `d=−log|C|`**：`|C|` が無限小に近づく零点（θ=π/4 近傍）で `d` は unlimited になる（nb08 S3）。
  → `d` は `\*R` 上で**有限とは限らない**内部関数。`st(d)` は零点で未定義（nb08 が明示）。
- **MDS（多次元尺度構成）**：標準では固有値分解（連続スペクトル）。`\*R` 上では超有限次元行列の
  内部固有値分解で、`st` を取って標準スペクトルに落とす。→ **pre-st は内部、post-st が標準幾何**。
- **層の判定**：MDS は **pre-st（内部）で構成し、st で標準幾何 S¹ を取り出す**。Q10 の円はこの post-st 像。

```python
constructs.append(Construct(
    name="距離核 d=-log|C|",
    layer=PRE_ST,
    star_object="内部関数 *R→*R∪{unlimited}。零点 θ=π/4 で unlimited",
    st_role="st(d) は零点で未定義（nb08 S3）。再スケール st(d/log(1/ε)) で有限座標。",
    note="飽和の議論では零点の扱いが効く。"))

constructs.append(Construct(
    name="MDS による空間復元",
    layer=PRE_ST,
    star_object="超有限次元内部行列の内部固有値分解",
    st_role="st でスペクトルを標準化し、標準幾何 S¹（=Q10 の円）を取り出す。",
    note="Q10 の『円 S¹』は MDS の post-st 像。循環判定はこの構成に懸かる。"))
```

### 2.3 測定 st と被覆 II

- **測定 `M(f)=st(f/ε^{ord(f)})`**：**st が発動する特権的操作**（nb08）。無限小オーダーで割ってから
  標準部を取る。これが pre-st → post-st の**主要な境界**。A-8（st は次元を縛らない）はこの操作の性質。
- **被覆 II（nb13）**：S¹ の普遍被覆 R への持ち上げ。位相的操作で、`\*R` 上では `\*S¹` の内部普遍被覆
  `\*R`（被覆座標 t）への持ち上げに移行。時間 R は `st` を取った標準被覆座標。
- **層の判定**：測定 st が境界。被覆 II は pre-st の位相操作（`\*S¹→\*R`）で、時間の標準化に st。

```python
constructs.append(Construct(
    name="測定 M(f)=st(f/ε^ord)",
    layer="boundary (pre→post)",
    star_object="st 発動の特権的操作。無限小オーダーを割って標準部を取る",
    st_role="★これが pre-st → post-st の主要境界。A-8: st は次元を縛らない。",
    note="nb08 で確立。E-2 の飽和が st の前か後かを §5 で確定する。"))

constructs.append(Construct(
    name="被覆 II（時間創発）",
    layer=PRE_ST,
    star_object="*S¹ の内部普遍被覆 *R（被覆座標 t）への持ち上げ",
    st_role="標準時間 R = st(被覆座標)。位相操作自体は st 前。",
    note="nb13。時間の軸。nb30 Q11 で飽和とは独立と判定済み。"))
```

## 3. 微分・連続・極限の取り扱い規約（本ノートの主目的その一）

方針 (ii) に従い、E-2 で現れる微分・連続・極限を `\*R` 上の無限小演算で定義する。
**標準の ε-δ 極限論を使わず**、以下の規約で統一する。

**規約 D-1（微分＝無限小差分商）.** 内部関数 `f: \*R→\*R` の点 θ（標準）での微分を
`f'(θ) := st( (f(θ+δ) − f(θ)) / δ )`（δ は任意の 0 でない無限小、結果が δ の取り方に依らないとき）と
定義する。標準的な導関数と一致する（移行原理による）。**飽和を極値条件 `f'=0` で書くときはこの規約を使う**。

**規約 D-2（連続＝無限小近接の保存）.** `f` が θ（標準）で連続 ⇔ `θ'≈θ ⇒ f(θ')≈f(θ)`
（無限に近い入力が無限に近い出力を与える）。cos2θ への連続接続（L-c）はこの判定で書く。

**規約 D-3（極限＝標準部）.** 数列・関数の極限を `lim = st(超実数値)` で書く。連続極限（格子間隔
`ε→0`）は「ε を無限小に取り st を施す」操作に一本化する（nb08 の超有限格子の連続極限と同じ）。

**規約 D-4（積分＝超有限和の標準部）.** 連続積分を超有限和 `Σ` の `st` で定義する（Loeb 測度の精神）。
飽和を「相関集合上の最大化」で書くときの和・積分はこの規約。

**含意（E-2 で実際に効く点）**：
- 飽和を**極値問題**として書くなら D-1（`f'=0` を無限小差分商で）。
- 飽和を**凸境界**として書くなら D-2・D-4（境界の連続性・面積分を無限小演算で）。
- **Q7（局所直交性 / Lovász ϑ）は微分を使わず代数・組合せ的**（グラフのベクトル表現の内積）なので、
  D-1〜D-4 をほぼ要しない——これが Q7 の「微分方程式の取り扱い問題を回避できる」強み（nb30 §5 で
  HOLD 上層とした理由の一つ）。方針 (ii) の下でも Q7 は安全な後退先になる。

## 5. 内部 vs 外部の罠：飽和の達成（最大値の存在）をどう言うか

**問題.** 標準解析では「コンパクト集合上の連続関数は最大値を達成する」で飽和（Tsirelson 値の達成）を
言える。NSA（方針 ii）では、この**コンパクト性が内部性・超有限性に置き換わる**。ここを規約で固めて
おかないと、Q9/Q10 後半の「飽和を達成する状態が存在する」が宙に浮く。

**規約 E-1（飽和の達成は内部集合の上で言う）.** 「すべての相関にわたる最大化」で飽和を論じるとき、
最大化の対象集合は**内部集合**でなければならない。内部集合なら超有限性により最大値（または超実数値の
上限の達成元）が内部的に存在し、`st` を取って標準の飽和値を得る。**対象が外部集合（例：標準状態の集合を
\*R 上で外から眺めたもの）だと最大値の存在が保証されない**——この取り違えが NSA 特有の罠。

**規約 E-2（cos2θ への接続は D-2 で、達成は E-1 で）.** L-c（相関が cos2θ に乗る）は連続性の主張＝D-2、
L-b の達成（飽和点が実現される）は最大値の存在＝E-1、と**別規約で扱う**。両者を混ぜない。

**E-2 候補への含意**：
- **Q10（円の半径）**：最大化を経由せず円の半径として飽和が出るなら E-1 の罠を**回避**できる（強み）。
- **Q9（凸境界）**：相関集合の内部性を確認する必要（NPA 階層の各レベルは内部か？を要検討）。
- **Q7（Lovász ϑ）**：ϑ は有限グラフの量で内部性問題が**生じない**（さらなる強み）。

→ この §5 は nb31（Q10）以降で「飽和の存在証明」を書くときの**前提規約**になる。

### 教訓（D への追記候補）

- **D-34（候補）：基礎の舞台（標準 / NSA）を、計算に入る前に宣言せよ。** 微分・連続・極限の意味は
   舞台で変わる。舞台を曖昧にしたまま飽和や極値を論じると、循環や最大値の存在の罠に落ちる。
   舞台の宣言は rubric の宣言（D-29）と同格の「先に固定すべき前提」である。

- **D-35（候補）：NSA では「内部か外部か」が「コンパクトか否か」を置き換える。** 標準解析の
   コンパクト性に依存する存在証明（最大値・飽和の達成）は、NSA では内部性・超有限性に翻訳して
   初めて有効。翻訳を怠ると存在が宙に浮く。飽和を論じる全候補（Q9/Q10後半）で要点検。
