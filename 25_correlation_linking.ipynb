## 25.0 セットアップと、nb22 から引き継ぐ確定事実

nb22（C-1）で確定した特徴写像をそのまま土台にする。改変しない。

- 核は厳密 rank2：`C = Σ_{k=1}^2 φ_k(θ_i)φ_k(θ_j)`、`φ(θ) = (cos2θ, sin2θ)/√(2N)`。
- 像は **半径 `1/√(2N)` の円 S¹**（定ノルム・周期π の一価閉曲線）。

ここで C8 の最初の障害が即座に見える：**1本の設定軸の特徴像は「1つの円」**である。
**絡み目（linking）は2つ以上の閉曲線の間でしか定義されない。** 1つの円には自己絡みはない
（自己絡みは framing/writhe であって linking number ではない）。

したがって C8 が成立するには「**第二の閉曲線がどこから来るか**」を内在的に答えねばならない。
これが CK-1（密輸検査）の本体になる。


```python

import numpy as np
N = 3

# nb22 の特徴写像（確定事実、改変しない）
def phi(theta, N=3):
    return np.stack([np.cos(2*theta), np.sin(2*theta)], axis=-1)/np.sqrt(2*N)

th = np.linspace(0, 2*np.pi, 400, endpoint=False)
img = phi(th, N)
center = img.mean(0)
rank = np.linalg.matrix_rank(img - center, tol=1e-9)
radius = np.linalg.norm(img - center, axis=1).mean()
print(f"単一設定軸の特徴像： 仿射スパン次元 = {rank}（=2 平面、曲線は S¹）")
print(f"                    半径 = {radius:.5f}  （解析 1/√(2N) = {1/np.sqrt(2*N):.5f}）")
print()
print("帰結：1本の設定軸 → 1つの円 → 絡みは定義不能。")
print("C8 が立つには『第二の閉曲線』の内在的起源が必須（→ CK-1）。")
```

## 25.2 CK-2（一意性検査）：絡み数は d=2 で消え、d≥4 で自明化するか

C8 の生命線は **G5＝d を一意化する力**。絡み数（linking number）が d=3 でだけ非自明である、という
**位相の定理**を確認し、numerics で**例示**する（証明は数学であり numerics はその図示に留める＝D-20 の精神）。

- **d=2**：平面内の2つの素な単純閉曲線は、Jordan 曲線定理により一方が他方の**内側か外側**しかなく、
  絡み数という整数不変量が**存在しない**。
- **d=3**：1-サイクル同士に **ℤ 値の絡み数**（Gauss 積分）が定義され、Hopf link で ±1 と非自明。
- **d≥4**：1-サイクルは余次元 ≥3。一般の位置で2ループは**常にほどける**（絡み数は恒等的に0、不変量として自明）。

→ もし β または γ が「2つの 1-サイクル」を与えるなら、それらが非自明に絡みうる次元は**厳密に3**。
これは d を一意化する（G5 PASS）。numerics で Hopf link（d=3 で Lk=±1）と d≥4 相当の分離（Lk=0）を例示する。


```python

# Gauss 絡み数積分（離散版）。位相の定理の *図示* であり証明ではない。
def gauss_linking(c1, c2):
    lk = 0.0; n1=len(c1); n2=len(c2)
    for i in range(n1):
        dr1 = c1[(i+1)%n1]-c1[i]; r1=c1[i]
        for j in range(n2):
            dr2 = c2[(j+1)%n2]-c2[j]; d = r1 - c2[j]
            nd = np.linalg.norm(d)
            if nd < 1e-9: continue
            lk += np.dot(np.cross(dr1, dr2), d)/nd**3
    return lk/(4*np.pi)

t = np.linspace(0, 2*np.pi, 160, endpoint=False)
# d=3: Hopf link（2円が絡む）
A = np.stack([np.cos(t), np.sin(t), 0*t], 1)
B = np.stack([1+np.cos(t), 0*t, np.sin(t)], 1)
print(f"d=3  Hopf link            : Lk = {gauss_linking(A,B):+.3f}   （非自明 = ±1）")
# d>=4 の図示：4本目の座標で2円を分離すると絡みは消える（一般の位置でほどける）
A2 = np.stack([np.cos(t), np.sin(t), 0*t], 1)
B2 = np.stack([8+np.cos(t), 0*t, np.sin(t)], 1)
print(f"d≥4 相当（分離・ほどけ） : Lk = {gauss_linking(A2,B2):+.3f}   （自明 = 0）")
print()
print("d=2 は Lk 不変量が *存在しない*（Jordan）ため数値化の対象外＝定理として述べるのみ。")
print("→ 絡み数の非自明性は d=3 を一意化する（G5 PASS）。定理依拠で循環なし。")
```

## 25.3 CK-3（cos2θ 独立検査）：絡みは cos2θ 固有か、一般の閉曲線で立つか

E-1 は「`cos2θ` と**独立**な軸から次元3を登る」道だった（B-7, D-24）。
もし絡みの非自明性が **cos2θ という特定の関数形に依存**するなら、それは E-2（量子相関の上流）側へ
化けており（B-7 の警告）、E-1 の解にならない。逆に、絡みが「閉曲線でありさえすれば立つ」
**一般的性質**なら、C8 は cos2θ から正しく独立している。

検査：特徴像を cos2θ から、より弱い「等質な周期相関」（P0/A1 レベル）へ差し替えても、
得られる像が依然 S¹（閉曲線）であることを確認する。閉曲線であることは関数形の詳細に依らない。


```python

# cos2θ を、振幅・高調波を変えた一般の等質周期相関へ差し替えても像は閉曲線(S¹)のまま、を確認。
def feature_generic(theta, kind):
    if kind == "cos2θ (nb22)":
        v = np.stack([np.cos(2*theta), np.sin(2*theta)], -1)
    elif kind == "cosθ (P0等質・弱い)":
        v = np.stack([np.cos(theta), np.sin(theta)], -1)
    elif kind == "歪んだ周期 (高調波混入)":
        v = np.stack([np.cos(theta)+0.3*np.cos(3*theta),
                      np.sin(theta)+0.3*np.sin(3*theta)], -1)
    return v

th = np.linspace(0, 2*np.pi, 600, endpoint=False)
print("像が閉曲線(S¹)か：始点と終点の一致＋単一連結成分で判定")
for kind in ["cos2θ (nb22)", "cosθ (P0等質・弱い)", "歪んだ周期 (高調波混入)"]:
    v = feature_generic(th, kind)
    closed = np.allclose(feature_generic(np.array([0.0]), kind),
                         feature_generic(np.array([2*np.pi]), kind), atol=1e-9)
    print(f"  {kind:<22}: 閉曲線={closed}")
print()
print("→ どの等質周期相関でも像は閉曲線。絡みの土台(閉曲線性)は cos2θ 固有でない。")
print("  CK-3 PASS：C8 は cos2θ と独立（E-1 の正しい側）。ただし『独立すぎる』含意は 25.4 で。")
```

## 25.4 判定：C8 は「半分正しく半分外れる」（D-24）
