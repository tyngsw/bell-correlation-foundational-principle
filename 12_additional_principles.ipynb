## 12.1 追加原理の候補（4つ）と評価

| 候補 | 中身 | 最小性 | 両立性 | 目標適合 | 原理性 | 効く目標要素 |
|---|---|---|---|---|---|---|
| 減衰原理（道I の正当化）| 相関に有限相関長 ξ を課す | 低 | 中 | **高** | **最低** | 非コンパクト化 |
| 符号原理（道II）| sign(C) の時間方向を一つ選ぶ | **高** | **最高** | 中 | 中〜高 | 単一時間 |
| 動力学原理 | β,N をスケール依存に（くりこみ群）| 中 | 中 | 中 | **高** | 連続極限・次元 |
| 非二体相互作用 | 三体以上の相関でランク2を破る | 中 | 中 | **最高** | 中 | 非コンパクト化（根本）|

**重要な構図**：単独で目標を全部出せる候補はない（本丸 no-go の当然の帰結）。目標は独立な
複数要素（非コンパクト・単一時間・次元）を持つので、**役割の違う原理の組み合わせ**になる見込み。
符号原理・動力学原理は本丸の結果（周期性は破れない）と整合的（周期性を破ろうとしない）。
減衰原理・非二体は周期性の打破を正面から要求し「与件の外」に出る度合いが最大。

**探索順（本Chat指定）**：減衰原理 → 符号原理 → 動力学原理。非二体は「必ず確認すべき」として保持。

## 12.2 減衰原理の関門：相関長を与件の `N` だけで固定できるか

減衰原理は目標適合が最強だが原理性が最弱。弱点の核心は「**なぜ減衰するか／相関長 ξ がいくつか**を
与件から導けず、自由パラメータ ε（or ξ）を外から入れざるを得ない」こと（道I の罠）。

唯一の生存条件：**ξ が与件の `N` だけで強制される**こと（本丸で β*=4N が N から出力されたように）。
ξ=f(N) なら自由パラメータは消え原理になりうる。これを検証する。

**先に見通し**：`E(θ)=cos2θ/(2N)` で `N` は**振幅** 1/(2N) にのみ入り、角度構造（周期）には
入らない（nb02：k*=4 は cos2 の周期から出て N は定数項に吸収）。相関長は角度方向のスケール
なので、振幅パラメータ N から角度スケールを出すのは構造的に苦しい。3つの機構で確かめる。

```python
import numpy as np

# 機構1: 距離 d=-log|C| の角度依存に N は入るか
print("機構1（距離のスケールに N が入るか）:")
for N in [3, 5, 10]:
    phi = np.linspace(0.001, np.pi/2-0.001, 1000)
    d = -np.log(np.abs(np.cos(2*phi))/(2*N))
    d_centered = d - d.mean()   # MDS の中心化に対応
    print(f"  N={N}: d中心化後の標準偏差={d_centered.std():.6f}")
print("  → 全N同値。d(phi)=-log|cos2phi|+log(2N) で N は定数シフト、中心化で消える。")
print("    距離の角度依存に N は皆無 → 相関長に N は効かない。")
```

```python
# 機構2: 臨界点 beta*=4N が有限相関長を決めるか
print("機構2（臨界点 beta*=4N）:")
print("  beta*=4N は転移点。そこで xi=∞（発散）。beta≠beta* では xi 有限だが beta 依存。")
print("  → N は『どの beta で発散するか』を決めるが『有限の相関長の値』は決めない。")
print("    有限 xi を N で固定、にはならない。")
```

```python
# 機構3: 内部 N 自由度（QCD: SU(N) の N）が角度方向に減衰構造を生むか
print("機構3（内部 N 自由度の角度的寄与）:")
print("  素朴モデル: N 個の位相シフトした cos2 の和 sum_a cos2(theta - 2pi a/N)")
for N in [3, 5, 10]:
    th = np.linspace(0, 2*np.pi, 2000, endpoint=False)
    s = sum(np.cos(2*(th - 2*np.pi*a/N)) for a in range(N))
    amp = np.abs(np.fft.rfft(s))/len(th)
    print(f"  N={N}: k=2振幅={amp[2]:.4f}, 和の max/mean比={np.abs(s).max()/(np.abs(s).mean()+1e-9):.3f}")
print("  → 等間隔位相シフトした同一周波数の和は完全に打ち消し合う（恒等的に 0）。")
print("    内部色自由度を素朴に角度へ入れても減衰どころか相関が消滅。相関長を生まない。")
```
