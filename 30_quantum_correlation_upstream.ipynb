### E-2 と E-1 の決定的な違い（rubric 設計の出発点）

nb21 §21.7 / nb24 の注意書きが明示する通り、**E-1 の判定をそのまま流用してはならない**：

- **E-1（次元3）**：与件 `cos2θ` と**直交な独立軸**（軸II）。次元3は cos2θ の登攀のどの段にも現れず、
  cos2θ へ化ける候補（C5 相関多面体など）は E-1 では **KILL** だった。
- **E-2（量子相関）**：**軸I の本筋**。ここでは「cos2θ へ化ける」ことは**失敗ではなく正解**。
  Tsirelson 飽和を出す原理が cos2θ/(2N) の角度依存性を再現するのは、まさに求めるもの。
  → **G3（cos2θ 独立）は E-1 の関門だが E-2 では無効化／反転する**（後述 G3' に置換）。

この反転を rubric に正しく織り込むのが本ノート最大の設計上の注意点である。

## 0. 適用する鉄の教訓（判定の根拠規則）

nb24 同様、判定は気分でなく引き継ぎ書 C 節を**機械的に**適用する。E-2 で特に効くもの：

- **D-10**（最重要）：パラメータ先決め＋意味の後付けは偽の原理。本物は値を**予言**する。
  → Tsirelson 値 `2√2`（または飽和を与える `cos`）を**入力のどこかに忍ばせていないか**を最優先で疑う。
- **D-28**（本セッション昇格）：**「記述できる」と「選択/予言できる」を峻別せよ。** E-2 で最大の罠。
  量子論を *仮定すれば* Tsirelson が出るのは当たり前。問うのは「**より弱い原理から量子（飽和）を選べるか**」。
  GPT 再構成の多くは「量子論一般を**特徴づける**」が、それは飽和点を**選択**する力とは別物。
- **D-13 / D-17**：数・群の一致は原理でない。「`2√2` が出る数式がある」は導出でない。
- **D-15**：否定は素朴版/精密版を区別。「情報因果律で出る」を素朴に受けず、循環の有無で精密化。
- **D-18**：肯定・否定を同じ精度で。UNK が一つでも残れば HOLD（GO に盛らない）。
- **D-21**：自分の結論を弱める検証を率先（GO 候補にも反証チェックを併記）。
- **D-29**：固定 rubric を先に決めてから機械適用（事後正当化の禁止）。

**除外規則（E-2 固有）**：
- **A-8 は E-2 では除外規則として使わない。** `st`（標準部写像）は次元を縛らないが、量子性（飽和）は
  st より上流とも下流とも未確定。E-1 の「st 系は解にならない」をそのまま持ち込まないこと。

## 1. 判定対象を正確に固定する（最重要・D-10 の前提）

「A2 を導く」と言うとき、**何を出力とみなすか**を曖昧にすると D-28 の罠に落ちる。
出力を3層に分解し、本ノートが要求する水準を明示する：

| 層 | 内容 | E-2 が要求する水準 |
|---|---|---|
| **L-a** | 相関が**古典束を超える**（CHSH > 2、ベル破れ） | 与件 P0 に内在（nb21 段0）。**出力でない**。 |
| **L-b** | 相関が**超量子でない**（CHSH ≤ 2√2、PR box 排除＝Tsirelson 上界） | **本ノートの主目標**。ここを選ぶ原理を探す。 |
| **L-c** | 相関が**飽和を達成し**角度依存が `cos`（倍角で cos2θ）になる | L-b ＋ 連続回転族で飽和を達成する状態が存在。L-b が出れば cos は幾何的に従属。 |

**したがって本ノートの「A2 を出す」＝ L-b（Tsirelson 上界の選択）を、量子論を仮定せずに導く**こと。
L-c（cos の形）は L-b が出れば Tsirelson 飽和の幾何（特徴写像 φ=(cos2θ,sin2θ)/√(2N) の円, nb22 A-1）から
ほぼ従属する——この従属関係自体が established（A-1）なので、勝負は L-b に集約される。

**D-10 の具体化（E-2 版）**：候補が `2√2` や `cos` を**結論でなく前提**に置いていないか。
特に「量子論（Hilbert 空間・Born 則）を仮定」は L-b を**仮定している**に等しく、自動 FAIL（後述 G6）。

## 2. 判定 rubric（先に固定する — D-29）

nb24 の G1–G5 を E-2 用に**改訂**する。改訂点を明示（これが本ノートの方法論的核心）：

| 関門 | 名称 | 問い | FAIL 条件 |
|---|---|---|---|
| **G1** | **非循環（D-10）** | 出力 `2√2`/飽和が、入力・公理に値として埋め込まれていないか | 飽和値・cos を仮定して飽和を出している |
| **G2** | **量子論非仮定（D-28・新設の核）** | Hilbert 空間/Born 則/C\*代数を**仮定していない**か | 量子形式を前提に置く（＝L-b を仮定） |
| **G3'** | **cos2θ 整合（E-1 から反転）** | cos2θ/(2N) の角度依存と**整合**するか（E-1 と逆向き） | 飽和を出すが cos と両立しない別形を強制 |
| **G4** | **群・代数水準（D-17）** | 「飽和」が本数合わせでなく**構造的必然**として出るか | 数合わせ（特定の数式の一致のみ） |
| **G5** | **予言力（D-10 正面）** | L-b を**一意に**選び、古典(2)も超量子(4)も**排除**するか | 飽和を**上界としてしか**与えず飽和の達成を選べない／他の値も同格 |
| **G6** | **minimality（E-2 固有・新設）** | 原理が cos2θ より**真に上流**（より弱い前提）か | cos2θ と同等以上を仮定（上流でない＝循環の変種） |

**改訂の論理**：
- **G2 を新設**：E-2 最大の失敗様式は「量子論を仮定して量子相関を出す」自己循環。これを独立関門に格上げ（D-28）。
- **G3 → G3'（反転）**：E-1 では「cos2θ を密輸したら FAIL」。E-2 では「cos2θ と整合しなければ FAIL」。
  方向が逆。E-1 の判定を機械的に流用すると正解候補を誤って KILL する（nb21 の警告の実装）。
- **G6 を新設**：再構成原理が「実は cos2θ と同じくらい強い仮定」なら上流原理になっていない（minimality 違反）。

**合否ラベル**（nb24 と同じ規約）：
- **GO** … G1–G6 を原理的に通過可能（FAIL が見えない）。計算で詰める価値あり。
- **HOLD** … 一部関門が UNK。先に概念を詰める。
- **KILL** … いずれかで FAIL。再挑戦は時間の無駄。

注意：GO は「正しい」でなく「**まだ落ちていない**」。D-21 に従い GO にも反証チェックを併記する。

## 2.1 セットアップ（rubric のコード化）

```python
# 候補原理を構造化データとして定義し、rubric G1-G6 を適用して go/hold/kill を機械判定する。
# 判定は「原理的に FAIL が見えるか」の事前評価であり、計算結果ではない（本ノートは安全装置）。

from dataclasses import dataclass, field
from typing import Dict, List

PASS, FAIL, UNK = "PASS", "FAIL", "UNK"  # UNK = 未決定（HOLD要因）
GATES = ("G1", "G2", "G3p", "G4", "G5", "G6")  # G3p = G3'（cos2θ 整合）

@dataclass
class Candidate:
    cid: str
    cls: str          # 分類 I-VI, 0
    name: str
    idea: str         # 一行アイデア
    gates: Dict[str, str] = field(default_factory=dict)
    note: str = ""

    def verdict(self) -> str:
        vals = [self.gates.get(g, UNK) for g in GATES]
        if FAIL in vals:
            return "KILL"
        if UNK in vals:
            return "HOLD"
        return "GO"

cands: List[Candidate] = []
print("rubric:", GATES)
```
