## 27.1 (L1)(L2) の確認：non-signaling と非可分性は rank で捉えられる

nb22 の特徴写像で、2体相関 `C(θa,θb)` の**階数**が可分性を捉える：
- **rank1** ⟺ `C=f(θa)g(θb)`（積＝可分＝古典分離可能）。
- **rank≥2** ⟺ 積に分解できない（非可分＝Bell 違反を許す）。

これを対照相関で確認する。重要：**これは代数（線型代数）の事実**であり、**どの次元でも**成り立つ。
ここまでに**位相も d=3 も登場しない**。


````python

import numpy as np
N=3; m=120
ta=np.linspace(0,2*np.pi,m,endpoint=False)
tb=np.linspace(0,2*np.pi,m,endpoint=False)
TA,TB=np.meshgrid(ta,tb,indexing='ij')

C_sep = np.cos(2*TA)*np.cos(2*TB)      # 可分（積）：rank1
C_ent = np.cos(2*(TA-TB))/(2*N)        # 非可分（nb22 qutrit）：rank2

def numrank(M,tol=1e-9):
    s=np.linalg.svd(M,compute_uv=False); return int((s>tol*s[0]).sum())

print(f"rank(C_sep) = {numrank(C_sep)}  （可分＝積＝古典分離可能）")
print(f"rank(C_ent) = {numrank(C_ent)}  （非可分＝Bell 違反を許す, nb22）")
print()
print("→ 非可分性（=Bell 違反の余地）は *階数 rank* という線型代数量で完全に捉えられる。")
print("  ここまで位相も次元 d=3 も一切現れない。(L2) は代数的事実。")
````

## 27.2 (L3) の本丸：非可分性は「位相的 linking」として現れるか

(L3) を検査する。non-signaling／非可分性が**真の位相的絡み**を強制するなら、
分離可能相関（rank1）は linking=0、非分離相関（rank2）は linking≠0 を、**埋め込みの取り方に依らず**
（位相不変量として）示すはずである。

**最初の障害（nb27 scratch で判明）.** 設定円 Alice `{φ(θa)}` と Bob `{φ(θb)}` は
**同一の S¹**（同じ特徴空間の同じ円）。同一の円に linking はない。だから「Alice 円と Bob 円」では
2つの異なるループにならない。

**次の自然な構成.** 結合写像 `Φ(θa,θb)=(φ(θa),φ(θb)) ∈ R²×R²=R⁴` を考え、設定トーラス T² の
2つの生成ループ（a-サイクル：θb 固定で θa を回す／b-サイクル：θa 固定で θb を回す）の像の linking を測る。


````python

def phi(t,N=3): return np.array([np.cos(2*t),np.sin(2*t)])/np.sqrt(2*N)

def gauss_linking(c1,c2):
    lk=0.0;n1=len(c1);n2=len(c2)
    for i in range(n1):
        dr1=c1[(i+1)%n1]-c1[i];r1=c1[i]
        for j in range(n2):
            dr2=c2[(j+1)%n2]-c2[j];d=r1-c2[j];nd=np.linalg.norm(d)
            if nd<1e-9:continue
            lk+=np.dot(np.cross(dr1,dr2),d)/nd**3
    return lk/(4*np.pi)

t=np.linspace(0,2*np.pi,80,endpoint=False)
# 自然な埋め込みは R⁴。1-サイクルは R⁴ で余次元2…2本の円は R⁴ で一般に unlink（linking 自明）。
# linking を測るには 3 座標へ射影する必要がある。射影の取り方を変えて安定な整数になるか試す。
print("結合写像 Φ:T²→R⁴ の a/b サイクル像の linking（3座標射影を変えて検査）：")
for cols in [(0,1,2),(0,1,3),(0,2,3),(1,2,3)]:
    acyc=np.array([np.concatenate([phi(s),phi(0.3)]) for s in t])[:,cols]
    bcyc=np.array([np.concatenate([phi(0.3),phi(s)]) for s in t])[:,cols]
    print(f"  射影 {cols}: Lk = {gauss_linking(acyc,bcyc):+.3f}")
print()
print("結果：値は射影ごとにバラつき、整数にならない。")
print("→ これは *位相不変量* ではない（射影依存・非整数）。真の linking number ではない。")
print("  自然な埋め込み次元は R⁴ であり、そこで2サイクルは unlink（nb25 CK-2 の d≥4 自明と整合）。")
````

## 27.3 (L3) は成り立たない：非可分性は「代数（rank）」であって「位相（linking）」ではない

検査結果を統合する（肯定・否定を同じ精度で、D-18）。

- **27.1**：non-signaling／非可分性は **rank** という線型代数量で**完全に**捉えられる。位相も d=3 も不要。
- **27.2**：その非可分性を**位相的 linking** として書こうとすると、(i) 設定円は同一 S¹ で2ループにならず、
  (ii) 結合写像の自然な埋め込みは R⁴ で、そこでは2サイクルが unlink、(iii) 3次元へ射影して測った
  「linking もどき」は**射影依存・非整数**で**位相不変量でない**。

**∴ (L3) は破れている。** 「絡み（linking）」は非可分性の**比喩**であって、非可分性が要求する
**本来の数学構造は rank（代数）**である。rank はどの次元でも定義でき、**d=3 を選ばない**。
non-signaling／因果律は「相関を rank≥2 にする」ことは要求するが、「2つの 1-サイクルを d=3 で絡める」
ことは**要求しない**。

これは CK-1 の亡霊が正体を現したということ：nb25 の `PASS*`（第二ループは内在的に取れる）は、
**「取れるループは linking を担わない」**という意味で甘かった。non-signaling が供給するのは
**第二の代数モード（rank の第2成分）**であって、**第二の位相ループ**ではない。
