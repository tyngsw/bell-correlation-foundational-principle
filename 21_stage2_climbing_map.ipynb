**最小原理 P0**：「ベル不等式を破る相関 C(a,b) が存在する」。これ以外何も仮定しない。
**登り先**：第一段階の所与 `E(θ) = cos2θ/(2N)`、N=3。

各段で「P0 から**強制**されるか／**追加措定**が要るか」を判定し、第一段階の地図と対応させる。

## 21.0 セットアップ

```python
import numpy as np
import sympy as sp
np.set_printoptions(precision=5, suppress=True, linewidth=120)

# CHSH 量：S = E(a,b) - E(a,b') + E(a',b) + E(a',b'), |S|>2 でベル破れ
def chsh(Efunc, a=0, ap=np.pi/4, b=np.pi/8, bp=3*np.pi/8):
    return abs(Efunc(a, b) - Efunc(a, bp) + Efunc(ap, b) + Efunc(ap, bp))

print("setup done. Tsirelson bound 2√2 =", round(2*np.sqrt(2), 4))
```

## 21.1 段0→段1：相関の存在 → 測定設定対の関数 C(a,b)

最小原理 P0 はベルの枠組み（2者 Alice/Bob、各々測定設定 a,b）を前提する。相関が設定対の関数
`C(a,b)` であることは P0 に内在し、追加措定を要しない。

```python
print("最小原理 P0 = ベルの枠組み（2者・各測定設定）。C = C(a,b) は P0 に内在。")
print("→ 段1『C は測定設定対の関数』は P0 から強制（追加措定なし）。")
```

## 21.2 段1→段2：設定対の関数 → 相対角のみの関数 C(a−b)

cos2θ は θ=a−b（相対角）のみの関数で、これは設定空間の**並進対称（等質性）**を意味する。P0 は
等質性を要求しない（非等質な相関もベルを破りうる）ので、追加措定 **A1: 設定空間の等質性** が要る。
反証可能チェック：非等質相関でも CHSH>2（ベル破れ）が起こることを数値で確認する。

```python
E_homog = lambda a, b: np.cos(2*(a-b))                                  # 等質（相対角のみ）
E_inhom = lambda a, b: np.cos(2*a)*np.cos(2*b) + 0.5*np.sin(2*a)*np.sin(2*b)  # 非等質
print(f"CHSH：等質相関 S = {chsh(E_homog):.3f}, 非等質相関 S = {chsh(E_inhom):.3f}")
print("→ 非等質でも S>2 がありうる。等質性はベル破れの必要条件でない。")
print("  段2『相対角依存』は追加措定【A1: 設定空間の等質性】が必要。")
```

## 21.3 段2→段3：相対角の関数 → cos の倍角 cos2θ

なぜ cos で、なぜ倍角（周期π）か。量子相関が CHSH の **Tsirelson 束 2√2 を飽和**する相関は
`E = cos(相対角)` に固定される。cos2θ がこれを厳密飽和することを確認する。追加措定 **A2: 量子
相関（Tsirelson 飽和＝Hilbert 空間構造）**。

```python
S_cos = chsh(lambda a, b: np.cos(2*(a-b)))
print(f"cos2θ 相関の CHSH = {S_cos:.4f},  Tsirelson 束 2√2 = {2*np.sqrt(2):.4f}")
print("→ cos 形は Tsirelson を厳密飽和。『量子相関の最大破れ』を措定すれば cos が出る。")
print("  段3『cos の倍角形』は追加措定【A2: 量子相関 = Hilbert 空間構造】が必要。")
print("  （倍角『2』は spin の幾何＝射影測度の二価性から。A2 に含まれる。）")
```

## 21.4 段3→段4：cos2θ → 振幅 1/(2N), N=3

振幅の `1/(2N)` と `N=3` は、README が明言する通り**導出でなく QCD 同定**。第一段階と同じく
追加措定 **A3: N=3 の QCD 同定**（不変）。

```python
print("1/(2N) の N=3 は QCD 同定（README：導出でなく同定）。")
print("→ 段4は追加措定【A3: N=3 の QCD 同定】。第一段階と同じ（不変）。")
```

## 21.5 次元3はどの段に現れるか（最重要）

核心の問い：測定設定 a,b はどんな空間の点か。その空間の次元（＝物理空間の次元3）は、cos2θ への
登攀のどの段で決まるか。

```python
print("測定設定 a,b の住む空間の次元は、登攀のどの段で決まるか？")
print()
print("段1–4 はすべて相対角 θ = a−b（1スカラー）だけで閉じる：")
print("  段1: C(a,b)        — 設定対（次元は未指定）")
print("  段2: C(a−b)        — 相対角1パラメータ")
print("  段3: cos2(a−b)     — 相対角1パラメータ")
print("  段4: cos2(a−b)/(2N) — 相対角1パラメータ")
print()
print("→ cos2θ の導出に測定空間の次元は一度も要らない（θ は2方向の相対角＝1スカラー）。")
print("  【重大】次元3は cos2θ の登攀のどの段にも現れない。")
print("  次元3は『cos2θ の上流』ではなく『cos2θ と直交な独立軸』だった。")
```

## 21.6 登攀地図の確定：二本の独立な軸

21.1〜21.5 から、第二段階の構造は**二本の独立な軸**に分かれることが分かる。

```python
print("="*64)
print("第二段階の登攀地図（最小原理 P0 から）")
print("="*64)
print()
print("【軸I：cos2θ への登攀】（相対角 θ の1スカラーで閉じる）")
print("  P0 ──強制──→ 段1 C(a,b)")
print("     ──A1（等質性）──→ 段2 C(a−b)")
print("     ──A2（量子=Tsirelson飽和）──→ 段3 cos2θ")
print("     ──A3（N=3 QCD同定）──→ 段4 cos2θ/(2N)")
print()
print("【軸II：物理空間の次元・位相】（軸I と直交、cos2θ に現れない）")
print("  測定空間の次元 → 次元3（nb17: 外部入力, 軸数=独立測定方向）")
print("  設定空間の位相 → 被覆II（nb13: S¹→R で時間, 与件の内側だが軸Iと別）")
print()
print("→ 第二段階は cos2θ を A1·A2·A3 の3措定まで分解できる（軸I）。")
print("  だが次元3（軸II）は軸I のどの措定にも含まれない独立問題。")
print("  『先に第二段階をやれば次元3が見える』は半分正しく半分外れ：")
print("  見えたのは『次元3が cos2θ と構造的に独立』という事実だった。")
```
