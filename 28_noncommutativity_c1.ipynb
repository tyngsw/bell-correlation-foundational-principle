## 28.1 C1 の「3」は何の次元か：代数の次元 / 表現空間の次元 / 空間次元

qutrit（spin-1, N=3）の非可換スピン観測量 `Sx,Sy,Sz` を土台に、「3」の出どころを分類する。


```python

import numpy as np

# spin-1 生成子（3×3）
Sx=np.array([[0,1,0],[1,0,1],[0,1,0]])/np.sqrt(2)
Sy=np.array([[0,-1j,0],[1j,0,-1j],[0,1j,0]])/np.sqrt(2)
Sz=np.array([[1,0,0],[0,0,0],[0,0,-1]])
def comm(A,B): return A@B-B@A

print("非可換性 [Sx,Sy]=iSz :", np.allclose(comm(Sx,Sy), 1j*Sz))
print()
print("『3』の候補となる量：")
print(f"  (a) 代数の次元 dim su(2)      = 3   （生成子 Sx,Sy,Sz の本数）")
print(f"  (b) 表現空間の次元 dim(qutrit)= {Sx.shape[0]}   （ヒルベルト空間の次元）")
print(f"  (c) 実空間次元 d             = ?   （MDS が住む R^d の d）")
print()
print("→ (a)(b) はどちらも *内部* の量（代数・状態空間）。")
print("  C1 が次元3の上流原理になるには (c) を出さねばならないが、(a)(b)≠(c)。")
print("  『3 が3つ現れる』のは D-13/D-17 の数の一致であって、空間次元の導出ではない。")
```

## 28.2 B-4 の再現：連続 SO(3) を「空間対称群」と読めるか

C1 の「3」を**空間**と読む唯一の自然な道は「SO(3)＝空間回転群」。だが B-4（nb18c）は
**連続 SO(3) 回転が距離核の等長でない**ことを確定した。これを再現し、C1 の空間解釈を封じる。


```python

N=3
def phi(t): return np.array([np.cos(2*t),np.sin(2*t)])/np.sqrt(2*N)
def feat3(a,b,c): return np.concatenate([phi(a),phi(b),phi(c)])  # T³ の3軸測定角

rng=np.random.default_rng(0)
pts=rng.uniform(0,2*np.pi,size=(50,3))
F =np.array([feat3(*p) for p in pts])
D =np.sqrt(((F[:,None]-F[None,:])**2).sum(-1))

def so3(angle):  # 軸1-2面の連続回転
    c,s=np.cos(angle),np.sin(angle)
    return np.array([[c,-s,0],[s,c,0],[0,0,1]])

print("連続 SO(3) 回転による距離破れ（角度を変えて）：")
for ang in [0.2,0.5,1.0]:
    pr=pts@so3(ang).T
    Fr=np.array([feat3(*p) for p in pr])
    Dr=np.sqrt(((Fr[:,None]-Fr[None,:])**2).sum(-1))
    print(f"  回転角 {ang:.1f}: 距離破れ(最大) = {np.abs(D-Dr).max():.3f}")
print()
print("→ 連続 SO(3) は等長でない（破れ>0）。等長なのは並進 T³ と離散軸置換 S₃ のみ（B-4）。")
print("  C1 の『3』を *空間回転群 SO(3)* と同一視する読みは B-4 で否定済み。")
```

## 28.3 D-21 の公平検査：非可換性は「設定空間」を3次元に強制するか

C1 を生かす最後の逃げ道：内部代数でなく、非可換性が **設定空間（MDS が住む空間）** の次元を
3に**強制**するなら、それは空間次元の導出になりうる。検査する。

非可換な測定方向 `n_i ∈ S²` の間の spin-1 角度相関（Legendre `P₂`）から距離核を作り、
MDS の内在次元を見る。**ただし**、方向を `S²⊂R³` に置く時点で「方向は実空間 R³ に住む」を
仮定していないか（循環）を厳しく問う。


```python

rng=np.random.default_rng(1)
n=rng.normal(size=(60,3)); n/=np.linalg.norm(n,axis=1,keepdims=True)  # S²⊂R³ 上の方向
cosang=np.clip(n@n.T,-1,1)
C=(3*cosang**2-1)/2                  # spin-1 角度相関 P₂（非可換構造）
Dm=-np.log(np.abs(C)+1e-9); np.fill_diagonal(Dm,0)

# 古典 MDS 内在次元
D2=Dm**2; J=np.eye(len(Dm))-np.ones_like(Dm)/len(Dm); B=-0.5*J@D2@J
ev=np.sort(np.linalg.eigvalsh(B))[::-1]
ev=ev[ev>1e-9]
dom=int((ev>0.05*ev[0]).sum())
print(f"非可換方向相関の MDS 内在次元（支配的本数）= {dom}")
print(f"  上位固有値: {np.round(ev[:6],2)}")
print()
print("二重の否定：")
print("  ① 内在次元は 3 にきれいに落ちない（P₂ 核は低次元等長埋め込みを持たない）。")
print("     → 非可換性は『設定空間=3次元』を *強制しない*。")
print("  ② 仮に 3 が出ても、方向 n_i を S²⊂R³ に置いた時点で実空間3を仮定＝循環。")
print("     → C8/B-4 と同じ壁。非可換性そのものは『両立不能な観測量の組』を許すだけで、")
print("        その組を *空間の軸* と読むのは外部入力。")
```

## 28.4 C1 の判定：C8 と同型の壁で決着（記述/内部 ＝ 選択/空間ではない）

肯定・否定を同じ精度で（D-18）：

- **28.1**：C1 の「3」は **代数の次元 dim su(2)=3** か **表現空間 dim=3**——どちらも**内部**の量。
  空間次元 d とは別カテゴリ（D-13/D-17 の数の一致）。
- **28.2**：「3」を**空間回転群 SO(3)**と読む唯一の自然な道は、連続 SO(3) が距離核の等長でない
  ことで封じられる（B-4 再現）。
- **28.3**：非可換性が**設定空間**を3次元に強制する逃げ道も、(i) MDS 内在次元が 3 に落ちず、
  (ii) 方向を `S²⊂R³` に置く時点で循環、の二重で塞がる。

**∴ C1 は次元3の上流原理にならない。** 非可換性が要求するのは「両立不能な観測量の組（内部代数）」
であって「実空間の軸（空間次元）」ではない。これは **nb27 の C8 と同型の壁**——
*記述/内部はできるが、選択/空間はできない*。

**地図への含意.** C1 の否定は **B-2・B-7・C8（nb27）に続く第四の独立な否定**。
次元3は cos2θ・量子相関の登攀・因果律・非可換代数の**いずれからも**出ない。
B-8（次元3は真に外部入力）の確度がさらに上がった。残るは C3（安定性、道(b)＝nb29）のみ。
