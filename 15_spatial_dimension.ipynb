# notebook 15 — 追加原理の探索 (4)：空間次元3を定める第二の原理

nb12〜nb14 で追加原理の4候補を評価し終え、周期性を破って目標 (3+1) の**時間・非コンパクト**
要素を出す源泉は**被覆原理 II（nb13）ただ一つ**に集約された。減衰原理は脱落、符号原理 I は
II に吸収、非二体は II の言い換え、動力学原理は周期性を破らず別役割（連続極限・次元）で共存。

残る独立な目標要素は**空間次元 3**である。被覆 II は時間（1）と非コンパクト（R）を出すが、
空間が 3 次元であることは出さない。この notebook は「空間次元 3 を定める第二の原理」を探す
（nb14 open-1、最優先）。

**既確立の足場**：
- nb04（A-4）：独立 p 設定 → トーラス T^p、次元は独立設定の本数。
- nb13（II）：1因子を時間 R に持ち上げ → R × T^(p-1)。空間は p−1 個の S¹。
- nb12 機構1：N は振幅 `1/(2N)` のみに入り、角度構造・次元構造には入らない。

**問い**：空間 3 次元（⟺ 独立設定本数 p=4）を**与件から強制**できるか。

**最重要の規律（D-10）**：「p=4 だと (3+1) になる」「空間3＝色3」と後から同定するのは
**偽の原理**（反証可能性を消す後付け）。p または空間次元 3 を**独立に決める機構**か、
**選ぶ余地のない強制**でなければ原理にならない。本notebookは安易な数の一致を徹底して退ける。

## 15.0 セットアップ

```python
import numpy as np
from numpy.linalg import eigh
np.set_printoptions(precision=4, suppress=True, linewidth=120)

N = 3  # 与件（QCD との同定）
print("setup done. N =", N)
```

## 15.1 機構1：N=3 から空間次元が決まるか

最も自然な期待は「空間3次元＝色 N=3」である。だが nb12 機構1 で「N は振幅 `1/(2N)` のみに
入り、角度構造・次元構造には入らない」ことが確定している。N と次元が独立であることを、
独立設定本数 p とトーラス次元の N 非依存性として確認する。

```python
print("独立設定本数 p とトーラス次元（N 依存性を見る）：")
for p in [1, 2, 3, 4]:
    # T^p の最大固有値の縮退度 = 2p（nb04, 埋め込み次元）。N は振幅のみ。
    print(f"  p={p}: 埋め込み次元 2p={2*p}, 内在次元 p={p}（N=3 でも 5 でも同じ、N 非依存）")
print()
print("→ 次元は p だけで決まり N と無関係（nb12 機構1 と整合）。")
print("  『空間3＝色3』は N を次元に密輸入する後付け（D-10 の罠）。機構1 は不可。")
```

## 15.2 機構2：次元のパリティ — 「2p 型」で奇数3が出るか

nb04 の「次元 2p」は**埋め込み次元**（各 S¹ を MDS で張るのに2座標要る）であって、
**内在次元**ではない。多様体としての内在次元は各 S¹ が 1、T^p で p。目標の「空間3次元」は
内在次元の話なので、奇数3はパリティ的に問題なく出る（空間 = 内在次元 p−1）。

```python
print("内在次元で整理（被覆 II で時間1因子を引いた残りが空間）：")
for p in [2, 3, 4, 5]:
    print(f"  p={p}: 時空内在次元 = p = {p} → 時間1 + 空間{p-1}")
print()
print("→ (3+1) 時空 ⟺ 独立設定本数 p=4（空間 p−1=3）。")
print("  空間3 は内在次元なので奇数パリティの問題はない（『2p型で偶数のみ』は埋込次元の誤解）。")
print("  焦点は『p=4 を何が強制するか』に絞られる。")
```

## 15.3 機構3：p=4 を強制する自己無撞着はあるか

p（独立な測定設定の本数）を 4 に強制する候補を検討する。とくに本丸の臨界点 `β*=4N=12`
（nb10）の係数「4」が時空次元4と関係するか——魅力的な一致だが、これこそ D-10 が警告する
偽の一致になりうる。係数の起源を厳密に追う。

```python
# β*=4N の係数4の起源（nb10）：1 = (β/2N)·⟨cos²2s⟩₀ = (β/2N)·(1/2) → β* = 2N/⟨cos²⟩ = 4N
mean_cos2 = 0.5  # ⟨cos²2s⟩₀ over uniform s
coef = 2.0 / mean_cos2
print(f"β*=4N の係数の起源：β* = 2N / ⟨cos²2s⟩₀ = 2N / {mean_cos2} = {coef:.0f}·N")
print(f"  係数 {coef:.0f} = 2 / ⟨cos²2s⟩ = 2 / (1/2) = 4。これは cos² の一様平均 1/2 から来る")
print(f"  統計係数であって、時空次元 4 ではない。")
print()
print("→ 『4N の 4』と『次元 4』を同一視するのは偽の一致（D-10）。")
print("  係数4 は ⟨cos²⟩=1/2 の代数的帰結。p=4 を強制する自己無撞着は素朴版では無い。")
print("  与件 cos2θ は1設定の相関であり、独立設定の本数 p は与件に含まれない（nb04 で p は入力）。")
```

## 15.4 機構4：測定方向空間 SO(3) が独立設定本数に上限を与えるか

qutrit（spin-1, N=3）のベル測定は、実空間 R³ の方向（Stern–Gerlach 軸など）で行われる。
測定方向の回転群は SO(3)（3パラメータ）。独立な直交測定軸は最大3本。これが独立設定本数 p に
構造を与えるかを検討する。色 SU(3)（8次元）ではなく、測定方向の SO(3) であることに注意。

```python
print("qutrit の測定方向空間：")
print("  色 SU(3) は8次元（Gell-Mann 行列）。だが測定『方向』を回すのは実空間回転 SO(3)。")
print("  ベル測定の角度 θ は SO(3) 回転の1パラメータ断面。独立設定 = 独立回転軸。")
print("  SO(3) の独立生成子（直交軸）は3 → 独立な空間設定の自然な本数は3。")
print()
print("  空間：SO(3) の3軸 → 各軸に角度 S¹ → T³（内在3次元）")
print("  時間：被覆 II（cos2θ の周期を R に展開）→ R（1次元）")
print("  → (3+1) = SO(3) 空間3 + 被覆時間1。空間3 の候補起源が得られた。")
print()
print("【D-10 自己点検】SO(3) の次元3 と ヒルベルト次元 N=3 は『どちらも3』だが意味が違う：")
print("  - N=3   : qutrit の内部自由度（色）。振幅 1/(2N) に入る。")
print("  - SO(3) : 測定方向を回す実空間回転の自由度。設定空間の構造。")
print("  両者の一致を『3だから同じ』と短絡してはならない。起源は別。")
```

## 15.5 機構5の厳密判定：SO(3) 起源は原理か、循環か

機構4の「空間3＝SO(3) の3軸」は魅力的だが、**循環の危険**がある：「空間が3次元なのは
（測定が行われる）実空間が3次元だから」では、説明したい当のもの（空間の3次元性）を前提に
使う同義反復になる。原理として成立するか、循環に過ぎないかを厳密に判定する。

```python
print("判定基準：測定方向空間の次元が、創発する空間次元と『独立に』決まっているか。")
print("  - 測定方向空間 = 創発空間そのもの なら循環（同じものを2度数えた）。")
print("  - 測定方向空間（実験設定）≠ 創発空間（相関出力）なら非循環。")
print()
print("nb04 の構造：独立 p 設定 → 内在 p 次元。これは『入力設定数 = 出力次元』。")
print("  つまり創発空間の次元は、もともと入力 p で決まっていた。")
print()
print("検証：与件 cos2θ から SO(3) の3軸が導けるか")
print("  cos2θ は2測定方向の相対角 θ の関数（1パラメータ）。")
print("  3軸の独立性（直交3方向）は cos2θ に含まれない情報。")
print("  → SO(3) の3 は与件 cos2θ から導けない。外部入力")
print("    （『測定が3次元実空間で行われる』という物理的事実）。")
print()
print("判定：機構5 は循環ではない（測定方向空間 ≠ 創発空間）が、")
print("  『実空間3次元』を前提に使う点で、空間次元3の"
      "『与件からの独立導出』にはなっていない。")
```

## 15.6 結論：次元3 を与件から導く原理は無い（否定的確定）。ただし起源分離の洞察

空間次元3の起源として、本枠組みが取りうる立場は2つに尽きる。

**立場(A)**：独立設定本数 p を入力として与える（nb04 の素直な立場）。p=4 なら (3+1)。
ただし p=4 は与件 cos2θ から導けない外部入力。「なぜ4か」には答えない正直な立場。

**立場(B)**：空間 p−1=3 を測定方向 SO(3) の3軸から、時間1 を被覆 II から取る。SO(3) の3 も
与件 cos2θ から導けず、「測定が3次元実空間で行われる」という外部事実。循環ではないが、
「実空間3次元」を前提に使う点で次元3の独立導出ではない。

**どちらも「空間次元3を与件から導く」ことには成功しない。これが誠実な結論。**

一方、立場(B) は重要な**構造的洞察**を与える：**時間と空間で起源が違う**。
- 時間 = 被覆 II（cos2θ の周期性を R に展開、**与件の位相構造**から）。
- 空間 = 測定方向の独立軸（SO(3)/実空間、**測定の物理的設定**から）。

この非対称性は物理的に自然である（時間と空間の起源が異なってよい）。被覆 II が「与件の内側」
（cos2θ の周期）から時間を出すのに対し、空間次元は「与件の外側」（測定がどんな空間で行われるか）
に依存する。**次元4 の原理化は依然として未解決**であり、これが残る最重要 open となる。
