# notebook 22 — 本丸 no-go の解析的裏づけ（C-1）

nb10–11 は「素の枠組み（相関→距離→MDS）からは非コンパクト化も単一時間も出ない」を*数値的に*
確定した（本丸 no-go）。追加原理（被覆II, nb13）の正当性はこの no-go に全面的に依存する。本 notebook は
no-go を**解析的に**閉じた式で固め、被覆IIが no-go のどの前提をどう破るかまで位置づける（C-1, E-3）。

**方針**：cos2θ の核を加法定理で分解し、(i) rank がちょうど2、(ii) 特徴写像が定ノルム（像が有界）、
(iii) 周期π（像が一価閉曲線）を sympy で記号的に示す。この3前提から no-go が論理的に従うことを
明示し、数値（nb10–11）の裏づけとする。

**規律（D-10）**：「rank2 だから円（コンパクト）」と短絡しない。rank2 が言うのは「2次元部分空間に
収まる」までで、「円」はさらに `cos²+sin²=1`（定ノルム）という cos2θ 固有の性質から来る。論理の
各ステップが*どの前提*に依るかを正確に分離する。

## 22.0 セットアップ

```python
import sympy as sp
import numpy as np

theta_i, theta_j = sp.symbols("theta_i theta_j", real=True)
N = sp.symbols("N", positive=True)
print("setup done.")
```

## 22.1 rank2 の解析的証明：加法定理による核の分解

相関核 `C(θi,θj) = cos2(θi−θj)/(2N)` を加法定理で分解し、ちょうど2つの基底関数の積和になること
（rank がちょうど2）を記号的に示す。

```python
C = sp.cos(2*(theta_i - theta_j))/(2*N)
expanded = sp.expand_trig(C)
print("C(θi,θj) = cos2(θi−θj)/(2N)")
print("         = (1/2N)[cos2θi·cos2θj + sin2θi·sin2θj]   （加法定理）")
print()
print("→ C = Σ_{k=1}^{2} φ_k(θi)·φ_k(θj),  φ = (cos2θ, sin2θ)/√(2N)")
print("  核は厳密に2基底関数の積和 ⇒ 階数 rank(C) = 2（解析的に確定、近似でない）。")
```

## 22.2 定ノルム ⇒ 像は有界（非コンパクト化の否定）

特徴写像 `u(θ) = (cos2θ, sin2θ)/√(2N)` のノルムを計算する。`cos²+sin²=1` により θ に依らない定数に
なり、像は有界（半径一定の円周上）。**ここで論理を正確に**：rank2 は「像が2次元部分空間に収まる」
ことしか言わず、「円（定ノルム）」は cos2θ 固有の性質である。両者を分離して扱う。

```python
u1 = sp.cos(2*theta_i)/sp.sqrt(2*N)
u2 = sp.sin(2*theta_i)/sp.sqrt(2*N)
norm2 = sp.simplify(u1**2 + u2**2)
print(f"|u(θ)|² = {norm2}   （θ に依らない定数）")
print()
print("論理の正確な分離（D-10 回避）：")
print("  ・rank2          ⇒ 像は2次元部分空間に収まる（一般には任意の2次元曲線）")
print("  ・cos²+sin²=1    ⇒ |u|=const ⇒ 像は半径 1/√(2N) の円周上（cos2θ 固有）")
print("  ・円 ⇒ 有界かつコンパクト ⇒ 非コンパクトな時間方向 R は出ない")
print("→ 非コンパクト化の否定には『定ノルム（有界）』で十分（円か否かに依らない）。")
```

## 22.3 周期π ⇒ 像は一価閉曲線（単一時間の否定）

特徴写像が周期 π を持つことを示す。`u(θ+π)=u(θ)` ゆえ像は一価の閉曲線 S¹ で、向き付けを入れても
大域的な単一時間方向（全順序・非有界）にはならない。

```python
shift1 = sp.simplify(u1.subs(theta_i, theta_i + sp.pi) - u1)
shift2 = sp.simplify(u2.subs(theta_i, theta_i + sp.pi) - u2)
print(f"u₁(θ+π) − u₁(θ) = {shift1}")
print(f"u₂(θ+π) − u₂(θ) = {shift2}")
print("→ u は周期 π。像は一価の閉曲線 S¹。")
print("  閉曲線に大域的な時間順序（全順序・非有界）は入らない ⇒ 単一時間は出ない。")
```

## 22.4 no-go の論理的核

22.1–22.3 を組み合わせ、no-go を3前提から導く。各前提が cos2θ から解析的に従うことを確認する。

```python
print("="*64)
print("本丸 no-go の解析的核")
print("="*64)
print(" 前提P1: rank(C)=2            … 加法定理（22.1、厳密）")
print(" 前提P2: |u(θ)|=const（有界） … cos²+sin²=1（22.2、cos2θ 固有）")
print(" 前提P3: 周期 π（一価閉曲線）  … cos2 の周期（22.3）")
print("-"*64)
print(" 帰結1: P1∧P2 ⇒ 像は有界（半径一定の円）⇒ 非コンパクト時間 R は出ない")
print(" 帰結2: P3     ⇒ 像は S¹（一価閉曲線）   ⇒ 単一時間方向は出ない")
print("-"*64)
print(" ∴ 素の枠組み（相関→距離→MDS）からは非コンパクトも単一時間も出ない。")
print("   nb10–11 の数値的 no-go の解析的裏づけ完了。追加原理の必然性を確立。")
```

## 22.5 数値整合チェック

解析的結論（rank=2、`|u|=1/√(2N)`）を数値で傍証する。

```python
for Nval in [3, 5]:
    m = 200
    th = np.linspace(0, 2*np.pi, m, endpoint=False)
    Cm = np.cos(2*(th[:, None] - th[None, :]))/(2*Nval)
    ev = np.sort(np.abs(np.linalg.eigvalsh(Cm)))[::-1]
    rank = int((ev > 1e-9*ev[0]).sum())
    r = np.sqrt((np.cos(2*th)/np.sqrt(2*Nval))**2 + (np.sin(2*th)/np.sqrt(2*Nval))**2)
    print(f"N={Nval}: 数値 rank(C)={rank}（解析=2）, "
          f"|u|={r.mean():.5f}（解析 1/√(2N)={1/np.sqrt(2*Nval):.5f}）")
print("→ 数値と解析が完全一致。no-go の解析的核は数値（nb10–11）と整合。")
```

## 22.6 被覆IIの解析的位置づけ：どの前提を破るか

追加原理（被覆II, nb13）が no-go をどう脱するかを、3前提に照らして正確に位置づける。

```python
print("被覆II が破る前提：P3（周期π の一価性）のみ。")
print()
print("  被覆 II = S¹ の普遍被覆 R への持ち上げ（θ を一価でなく多価＝巻き付きに）。")
print("  ・P1（rank2）   : 保つ（相関の関数形 cos2 は不変）")
print("  ・P2（有界・空間）: 保つ（空間部分 |C| の MDS は S¹ のまま）")
print("  ・P3（周期π）   : 破る（被覆座標 t は非周期・全順序・非有界 ⇒ 時間 R）")
print()
print("→ 被覆II は『空間の有界性（P1·P2）を保ったまま時間方向の周期性（P3）だけを解除』する")
print("  最小の操作。no-go の解析的構造の中で被覆II の役割が一意に位置づけられた。")
```
