## 22b.0 セットアップと記号規約

nb08 の超有限格子規約を引き継ぐ：無限小 `ε>0`（`ε≈0, ε≠0`）を固定し、超有限格子
`θ_i = i·ε`（i は \*Z の超整数、`0 ≤ i < M`、`M·ε = 2π` または基本領域幅）を設定空間の台とする。
`\*cos, \*sin` は移行で持ち上げた超実数値関数。`st` は有限超実数の標準部。

記号：標準対象に対する \*R 上の持ち上げを `\*` で表す（`\*S¹` 等）。コードは記号計算（sympy）で
**移行で成立する代数恒等式**を確認するのみ——超実数自体は数値化できないため、恒等式が θ の値に
依らず成立することをもって「移行で \*R 上にも成立」と読む（移行原理の計算的表現）。

```python
import sympy as sp

# 超実数 θ は記号として扱う（特定の標準値を代入しない＝移行の計算的表現）。
theta, theta_i, theta_j, delta = sp.symbols("theta theta_i theta_j delta", real=True)
N = sp.symbols("N", positive=True)
eps = sp.symbols("epsilon", positive=True)  # 無限小（記号）

print("nb22b setup: 超有限格子 θ_i = i·ε, 基礎体 *R。")
print("コードは『θ に依らず成立する代数恒等式』を確認し、移行原理で *R 上の成立と読む。")
```

## 22b.1 移行原理で \*R へ持ち上がる部分（P1・P2・P3 の自動持ち上げ）

nb22 の三前提はいずれも**一階の代数的命題**である。nb30b 規約 F-2（移行原理）により、R 上で
成立するこれらは \*R 上でもそのまま成立する。各命題が θ の特定値に依らない恒等式であることを
sympy で確認し、それをもって移行による \*R 上の成立とする。

### 22b.1.1 P1（rank2）の持ち上げ：加法定理は \*R 上の恒等式

```python
# 加法定理は体の恒等式 ⇒ 移行で *R 上にも成立（θ_i, θ_j が超実数でも）。
C = sp.cos(2*(theta_i - theta_j))/(2*N)
expanded = sp.expand_trig(C)
# φ=(cos2θ, sin2θ)/√(2N) の積和に一致するか
phi_form = (sp.cos(2*theta_i)*sp.cos(2*theta_j) + sp.sin(2*theta_i)*sp.sin(2*theta_j))/(2*N)
diff = sp.simplify(expanded - phi_form)
print("C(θi,θj) - Σ_k φ_k(θi)φ_k(θj) =", diff, " (恒等的に 0)")
print("⇒ 加法定理は体の恒等式。移行原理で *R 上でも C = Σ_{k=1}^2 φ_k φ_k。")
print("⇒ P1: rank(*C) = 2 が *R 上で成立（超実数 θ に対しても厳密に2基底）。")
```

### 22b.1.2 P2（定ノルム）の持ち上げ：cos²+sin²=1 は \*R 上の恒等式

```python
# 定ノルムは cos²+sin²=1（体の恒等式）から。移行で *R 上にも成立。
u1 = sp.cos(2*theta)/sp.sqrt(2*N)
u2 = sp.sin(2*theta)/sp.sqrt(2*N)
norm2 = sp.simplify(u1**2 + u2**2)
print(f"|*u(θ)|² = {norm2}   （θ に依らない＝恒等式）")
print("⇒ cos²+sin²=1 は体の恒等式。移行で *R 上でも |*u(θ)|² = 1/(2N)。")
print("⇒ P2: 像 *S¹ は半径 1/√(2N) の超実数円上（超実数 θ に対しても定ノルム）。")
```

### 22b.1.3 P3（周期π）の持ち上げ：周期性は \*R 上でも成立

```python
# 周期 π。u(θ+π)=u(θ) は移行で *R 上にも成立（超実数 θ に対しても）。
s1 = sp.simplify(u1.subs(theta, theta + sp.pi) - u1)
s2 = sp.simplify(u2.subs(theta, theta + sp.pi) - u2)
print(f"*u₁(θ+π) − *u₁(θ) = {s1}")
print(f"*u₂(θ+π) − *u₂(θ) = {s2}")
print("⇒ 周期 π は移行で *R 上にも成立。像は一価の内部閉曲線 *S¹。")
print()
print("【22b.1 結論】P1・P2・P3 はすべて一階の代数命題ゆえ移行原理で *R へ自動的に持ち上がる。")
print("  → no-go の論理的核（nb22 §22.4）は *R 上でそのまま成立する。")
```

## 22b.2 超有限格子上での明示的構成（移行の外：依存構造を見るため）

移行原理は命題の**真偽**を保存するが、命題が**何に依存して成り立つか**（証明の構造）は教えない。
Q10 の循環判定はこの依存構造を要求するので、特徴写像を超有限格子 `θ_i=i·ε` の上に**実際に構成**し、
円 \*S¹ の半径がどの量から決まるかを追跡する。

**構成.** nb08 と同一規約で `θ_i = i·ε`（`i ∈ \*Z, 0≤i<M`）。各格子点に特徴ベクトル
`\*u(θ_i) = (\*cos2θ_i, \*sin2θ_i)/√(2N) ∈ \*R²` を割り当てる。像 `\*S¹ = { \*u(θ_i) : 0≤i<M }`。

### 22b.2.1 内部性と半径（nb30b 規約 E-1 の前提固め）

```python
# 像 *S¹ は超有限集合 {*u(θ_i)} で、内部集合（超有限ゆえ自動的に internal）。
# 半径は θ_i に依らず 1/√(2N)（標準有限）。st が定義でき、st(*S¹) = 標準円 S¹。
print("像 *S¹ = { *u(θ_i) : 0≤i<M } は超有限集合 ⇒ 内部集合（internal）。")
print("  ・nb30b 規約 E-1（飽和の達成は内部集合上で言う）の前提を満たす。")
print("  ・半径 |*u(θ_i)| = 1/√(2N)（標準有限・θ_i 非依存）⇒ st 定義可能。")
print("  ・st(*S¹) = 標準円 S¹（半径 1/√(2N)）。")
print()
print("⇒ Q10 が論じる『円の半径』は、内部集合 *S¹ の標準有限な半径 1/√(2N)。")
print("  最大値の存在（飽和の達成）を内部集合上で言える土台が整った。")
```

### 22b.2.2 微分規約 D-1 の不要性（この構成では代数で閉じる）

```python
# 半径は |*u|²=1/(2N) という代数恒等式から出る。微分（無限小差分商 D-1）を使わない。
# 念のため、無限小差分商で接ベクトルを見ても半径（ノルム）は不変であることを確認。
du1 = sp.simplify((u1.subs(theta, theta+delta) - u1))   # 差分（standard part 前）
du2 = sp.simplify((u2.subs(theta, theta+delta) - u2))
# ノルムの差分商の極限ではなく、ノルム自体が定数であることが本質
print("特徴写像のノルム |*u|² =", sp.simplify(u1**2+u2**2), "（定数）")
print("⇒ 半径は代数恒等式 cos²+sin²=1 から。微分規約 D-1 は不要。")
print("  Q10 の循環判定は pre-st の代数層で実行できる（nb30b §4.1 の見通しを確認）。")
```

## 22b.3 P2 の依存構造の予備切り分け（nb31 への橋渡し・本ノートの核心）

Q10 循環判定の核心は次の問いだった：
> 定ノルム P2（`|u|²=1/(2N)`）は、(a) `cos²+sin²=1` という**与件の振幅と独立な恒等式**から立つのか、
> (b) 振幅 `1/(2N)` の**措定**から立つのか。

(a) なら飽和は no-go の前提に密輸でなく自然に埋まっており Q10 は非循環（established な再記述）。
(b) なら飽和値を仮定して飽和を出す循環で Q10 は KILL。これを超有限構成で見えるところまで前進させる。

```python
# P2 の値 1/(2N) を「方向部分」と「振幅部分」に分解して依存を可視化する。
# *u(θ) = R(N) · ê(θ),  ê(θ)=(cos2θ, sin2θ) は単位ベクトル, R(N)=1/√(2N) は振幅。
e1, e2 = sp.cos(2*theta), sp.sin(2*theta)
unit_norm2 = sp.simplify(e1**2 + e2**2)         # 方向部分のノルム²
amp = 1/sp.sqrt(2*N)                             # 振幅 R(N)
print("分解 *u(θ) = R(N)·ê(θ):")
print(f"  方向 ê=(cos2θ,sin2θ): |ê|² = {unit_norm2}  ← cos²+sin²=1（N に依らない恒等式）")
print(f"  振幅 R(N) = 1/√(2N): R² = {sp.simplify(amp**2)}  ← 与件 cos2θ/(2N) の振幅から")
print()
print("【依存構造】 |*u|² = R(N)²·|ê|² = (1/2N)·1")
print("  ・『円である（ノルムが θ に依らず一定）』 ⇐ |ê|²=1（恒等式・振幅と独立）  …(a)")
print("  ・『半径の値が 1/√(2N)』            ⇐ 振幅 R(N)（与件の振幅 1/(2N) から）  …(b)")
```

### 22b.3.1 切り分けの帰結（暫定・nb31 で確定）

上の分解は、P2 が**二つの異なる起源の積**であることを示す：

- **円であること（定ノルム性・θ非依存）** は `|ê|²=cos²+sin²=1`、すなわち**与件の振幅と独立な恒等式**
  から来る（起源 a）。これは cos2θ の振幅が 1/(2N) であろうと別の値であろうと成立する。
- **半径の値 1/√(2N)** は与件の振幅 `1/(2N)` から来る（起源 b）。

**Q10 の循環判定への含意（暫定）**：
- Q10 が「**相関が飽和の幾何（円）に乗る**」ことを主張するなら、それは起源 (a) に依拠し、**与件の振幅を
  仮定しない＝非循環**。「飽和の質的構造（円）は no-go の前提に既に埋まっている」という established な
  再記述になりうる（nb30 Q10 の楽観シナリオを支持）。
- Q10 が「**飽和の値そのもの（Tsirelson 値 2√2 に対応する特定の半径）**」を主張するなら、それは起源 (b)
  に依拠し、与件の振幅 1/(2N) を経由する。ここで「振幅 1/(2N) が cos2θ の与件から独立に正当化されるか」
  が次の循環判定の焦点になる（N=3 は QCD 同定＝A3 で外部入力なので、**振幅は与件内で完結せず A3 に依存**）。

**したがって暫定の見通し**：Q10 は「**飽和の幾何的構造（円）は非循環で established**、しかし**飽和の
数値（半径＝振幅）は A3（N=3 同定）に依存**」という**二層の判定**になる見込み。これは nb30 §1 の
L-b/L-c 分解（質的な飽和 vs 数値）と対応し、D-24（半分正しく半分外れる＝前進）の典型になりうる。
**確定は nb31 で**（本ノートは予備切り分けまで）。

## 22b.4 被覆 II の \*R 上での位置づけ（nb22 §22.6 の持ち上げ）

nb22 §22.6 で被覆 II は P3（周期π の一価性）のみを破り P1・P2 を保つと確定した。\*R 上では：

```python
print("被覆 II の *R 持ち上げ：")
print("  *S¹ の内部普遍被覆 = *R（被覆座標 t）, 被覆写像 *π: *R → *S¹")
print("  ・P1(rank2)・P2(定ノルム)  : 保つ（代数構造不変、22b.1）")
print("  ・P3(周期π)              : 破る（被覆座標 t は非周期・全順序・非有界 ⇒ 時間 *R）")
print("  ・標準時間 R = st(t)")
print()
print("⇒ nb22 §22.6 の結論は *R 上でそのまま成立。被覆 II は飽和と独立（nb30 Q11）。")
```
